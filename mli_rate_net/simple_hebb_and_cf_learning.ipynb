```python
def update_firing_rate(u,PF,CF,w,A,time_step):
    dudt = -A*u + dot(PF,w) + CF
    return u + time_step*dudt

def update_weights(w,w0,u,PF,L,N,time_step):
    dwdt = -L*PF*(w-w0) + N*PF*u
    return w + time_step*dwdt

def PF_rate(a,b,t):
    if a<=t<=b:
        return 1.
    else:
        return 0.
```

```python
def run_trial(u,w,t0,tf,time_step):
    u_hist, w_hist = [u], [w]
    for t in arange(t0,tf+time_step,time_step):
        PF = array([PF_rate(a1,b1,t),PF_rate(a2,b2,t)])
        CF = 1. if t == cf_t else 0.
        u_hist.append(update_firing_rate(u_hist[-1],PF,CF,w_hist[-1],A,time_step))
        w_hist.append(update_weights(w_hist[-1],w0,u_hist[-1],PF,L,N,time_step))
    return u_hist, w_hist
```

```python
a1,b1,a2,b2 = 1,3,0,2
cf_t = 3.
N_trials = 200
w0 = array([0.,0.])
t0,tf,time_step = 0,3,.1
A,L,N = 8.,.195,1.
```

```python
u = 0.
w = array([0.,0.])#w0.copy()
u_hist_trials = []
w_trials = []
for n in range(N_trials):
    u_hist,w_hist = run_trial(u,w,t0,tf,time_step)
    u_hist_trials.append(u_hist)
    w = w_hist[-1]
    w_trials.append(w)
    u = 0.
    
```
