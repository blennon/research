<h2>Experiments with MLI
    & PKJ ISIs<h2>

The goal of this experiment is to reciprically connect a network of MLI and PKJ neurons and observe the the change in ISI statistics given some set of synaptic weight values

Discussion: The mean and variance of spontaneous currents for MLI cells are the same.  The same holds for PKJ cells (but with different means and variances).  By connecting this network with MLI->MLI, MLI->PKJ and PKJ->MLI synapses, the mean firing rates of both MLI and PKJ cells decrease and the ISI variances increase, as observed in Hausser and Clark 1997.  The means are heavily affected by synaptic weights and number of synapses each neuron receives.  So, choosing the weights and number of synapses carefully one can achieve the target mean firing rate.  The variances of ISIs are more difficult.  Interestingly, the variance of MLI neurons increase substantially.  The coefficient of variation (CV) for MLI cells reaches ~.34 in this set up (~.40 in Hausser and Clark 1997 control).  Interestly, the ISI CV for PKJs stays rather low ~.23 here vs .49 in H&C 1997.  Additionaly, we see an increased asymmetry relative to GABA block in these simulations. Another interesting observation is that the PKJ->MLI synapses have a significant affect on the ISI variance in MLI cells: ISI variance without PKJ->MLI synapses falls to ~.26.

By adding PKJ -> PKJ synapses, the ISI variance of PKJs increases substantially.

<h2>ISI Histograms<h2>

```python
def isi_mean_and_std(monitor):
    '''
    compute the mean and variance of interspike intervals
    of a group of neurons
    '''
    isi = []
    for n_ind, times in monitor.spiketimes.iteritems():
        isi += list(diff(times)*1000)
    return mean(isi), var(isi)**.5

def simpleaxis(ax):
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.get_xaxis().tick_bottom()
    ax.get_yaxis().tick_left()
```

```python
fig = figure(figsize=(20,5))
ax = fig.add_subplot(121)
hist_plot(MISI_MLI, newfigure=False, color='w')
mew, std = isi_mean_and_std(MS_MLI)
#s = 'mean rate = %0.1f Hz\nmean ISI = %0.1f ms\nISI std = %0.1f ms\nCV = %0.2f' % (mean(MR_MLI.rate),mew,std,std/mew)
s = 'rate = %0.1f Hz\nCV = %0.2f' % (mean(MR_MLI.rate),std/mew)
annotate(s,(20,0),xytext=(95,MISI_MLI.count.max()*.7),xycoords='data',fontsize=24)
simpleaxis(ax)
tick_params(labelsize=20)
xlabel('ISI (ms)', fontsize=20)
ylabel('Count', fontsize=20)
#title('MLI ISI Histogram',fontsize=20)
xlim([0,200])


ax = fig.add_subplot(1,2,2)
hist_plot(MISI_PKJ, newfigure=False, color='k')
mew, std = isi_mean_and_std(MS_PKJ)
#s = 'mean rate = %0.1f Hz\nmean ISI = %0.1f ms\nISI std = %0.1f ms\nCV = %0.2f' % (mean(MR_PKJ.rate),mew,std,std/mew)
s = 'rate = %0.1f Hz\nCV = %0.2f' % (mean(MR_PKJ.rate),std/mew)
annotate(s,(20,0),xytext=(60,MISI_PKJ.count.max()*.7),xycoords='data',fontsize=24)
tick_params(labelsize=20)
simpleaxis(ax)
xlabel('ISI (ms)', fontsize=20)
ylabel('Count', fontsize=20)
#title('PKJ ISI Histogram',fontsize=20)
xlim([0,140])
```
