```python
import cPickle
sys.path.append('../../')
from neuron_models import *
from brian import *
```

```python
def load_data(in_dir):
    traces = cPickle.load(open(in_dir+'traces.pkl'))
    isis = cPickle.load(open(in_dir+'isis.pkl'))
    spikes = cPickle.load(open(in_dir+'spikes.pkl'))
    return traces, isis, spikes
```

```python
def draw_smooth_spikes(trace, ap = array([ 0.025,  .2,  0.7, 1.,  0.65, .05])):
    '''
    replaces the value of 0 in trace arrays with the waveform 'ap'
    '''
    new_trace = []
    last_v = None
    for v in trace:
        if v == 0:
            new_trace += list(ap*(0-last_v) + last_v)
            continue
        new_trace.append(v)
        last_v = v
    return array(new_trace)
```

```python
def plot_aligned_traces(traces, spikes, n=30, delay=12, ax=None, **plotoptions):
    if ax is None:
        ax = subplot(111)
    for i in range(n):
        trace = extract_two_spike_trace(traces[i], spikes[i], float(10*ms), float(.25*ms))
        trace = draw_smooth_spikes(trace)
        plot(.25*ms*arange(trace.shape[0])*1000,trace*1000,**plotoptions)
    simpleaxis(ax)
    if delay is not None:
        mli_spike_time = argmax(trace)*.25+delay
        ax.annotate('',xy=(mli_spike_time,-55),xytext=(mli_spike_time-delay/1.5,-45),arrowprops=dict(facecolor='k', shrink=0.05),fontsize=16, color='k')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    xlim([-5,60])
    ylim([-75,0])
    add_scalebar(ax)
    #xlabel('time (ms)',fontsize=20)
    #ylabel('membrane potential (mV)',fontsize=20)
    #xticks(fontsize=16)
    #yticks(fontsize=16)
    #title('Membrane Potential Traces',fontsize=22)
```
