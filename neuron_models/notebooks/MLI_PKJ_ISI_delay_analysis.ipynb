```python
import cPickle
sys.path.append('../../')
from neuron_models import *
from brian import *
```

```python
def load_data(in_dir):
    traces = cPickle.load(open(in_dir+'traces.pkl'))
    isis = cPickle.load(open(in_dir+'isis.pkl'))
    spikes = cPickle.load(open(in_dir+'spikes.pkl'))
    return traces, isis, spikes
```

<h2>Case #1: Without MLI input</h2>

```python
traces, isis, spikes = load_data('/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay/2013-10-18T14:17:57.662971/')
```

```python
figure(figsize=(8,6))
hist(isis)
print 'mean:', mean(isis), 'variance:', var(isis)

figure(figsize=(12,9))
for i in range(30):
    trace = extract_two_spike_trace(traces[i], spikes[i], float(10*ms), float(.25*ms))
    plot(trace,color='k')
```
