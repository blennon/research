```python
import cPickle
sys.path.append('../../')
from neuron_models import *
from brian import *
```

```python
def load_data(in_dir):
    traces = cPickle.load(open(in_dir+'traces.pkl'))
    isis = cPickle.load(open(in_dir+'isis.pkl'))
    spikes = cPickle.load(open(in_dir+'spikes.pkl'))
    return traces, isis, spikes
```

```python
def plot_aligned_traces(traces, spikes, n=30, delay=12):
    fig = figure(figsize=(12,9))
    ax = fig.add_subplot(111)
    for i in range(n):
        trace = extract_two_spike_trace(traces[i], spikes[i], float(10*ms), float(.25*ms))
        plot(.25*ms*arange(trace.shape[0])*1000,trace*1000,color='k')
    simpleaxis(ax)
    if delay is not None:
        mli_spike_time = argmax(trace)*.25+delay
        ax.annotate('MLI Spike',xy=(mli_spike_time,-58),xytext=(mli_spike_time-delay/1.5,-50),arrowprops=dict(facecolor='red', shrink=0.05),fontsize=18, color='red')
    xlabel('time (ms)',fontsize=20)
    ylabel('membrane potential (mV)',fontsize=20)
    title('Membrane Potential Traces',fontsize=22)
```

<h2>Case #1: Without MLI input</h2>

```python
traces, isis, spikes = load_data('/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay/2013-10-18T14:17:57.662971/')

# plot ISI histogram
fig = figure(figsize=(8,6))
ax = fig.add_subplot(111)
hist(array(isis)*1000, 10, color='k')
simpleaxis(ax)
xlabel('ISI (ms)', fontsize=20)
ylabel('Count', fontsize=20)
title('ISI Histogram',fontsize=22)
print 'mean:', mean(isis), 'std:', var(isis)**.5

plot_aligned_traces(traces, spikes, 30, None)
```
