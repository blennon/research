```python
import cPickle
sys.path.append('../../')
from neuron_models import *
from brian import *
```

```python
def load_data(in_dir):
    traces = cPickle.load(open(in_dir+'traces.pkl'))
    isis = cPickle.load(open(in_dir+'isis.pkl'))
    spikes = cPickle.load(open(in_dir+'spikes.pkl'))
    return traces, isis, spikes
```

```python
def plot_aligned_traces(traces, spikes, n=30, delay=12, ax=None):
    if ax is None:
        ax = subplot(111)
    for i in range(n):
        trace = extract_two_spike_trace(traces[i], spikes[i], float(10*ms), float(.25*ms))
        plot(.25*ms*arange(trace.shape[0])*1000,trace*1000,color='k')
    simpleaxis(ax)
    if delay is not None:
        mli_spike_time = argmax(trace)*.25+delay
        ax.annotate('',xy=(mli_spike_time,-55),xytext=(mli_spike_time-delay/1.5,-45),arrowprops=dict(facecolor='k', shrink=0.05),fontsize=16, color='k')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_visible(False)
    ax.spines['bottom'].set_visible(False)
    xlim([-5,60])
    ylim([-75,0])
    add_scalebar(ax)
    #xlabel('time (ms)',fontsize=20)
    #ylabel('membrane potential (mV)',fontsize=20)
    #xticks(fontsize=16)
    #yticks(fontsize=16)
    #title('Membrane Potential Traces',fontsize=22)
```

```python
# comparison to non FFI case
fig=figure(figsize=(12,10))
traces_control, isis_control, spikes_control = load_data('/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay/2013-10-18T14:17:57.662971/')
traces_2nS, isis_2nS, spikes_2nS = load_data('/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay/2013-10-18T14:08:00.728776/')#load_data('/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay/2013-10-18T16:43:41.129179//')

ax = fig.add_subplot(311)
plot_aligned_traces(traces_control, spikes_control, 30, None, ax)


ax = fig.add_subplot(312)
plot_aligned_traces(traces_2nS, spikes_2nS, 30, delay=12, ax=ax)

ax = fig.add_subplot(325)
ax.hist(array(isis_control)*1000,15,alpha=.5,color='k')
ax.hist(array(isis_2nS)*1000,15,alpha=.7,color='w')
simpleaxis(ax)
xticks(fontsize=16)
yticks(fontsize=16)
xlabel('ISI (ms)',fontsize=16)
ylabel('Count',fontsize=16)

ax = fig.add_subplot(326)
in_dir = '/home/bill/research/data/neuron_models/molecular_layer/MLI_PKJ_ISI_delay_rand_nS/2013-10-22T09:56:43.780197/'
isis = cPickle.load(open(in_dir+'isis.pkl'))
g_inhs = cPickle.load(open(in_dir+'g_inh.pkl'))
g_inh = array([float(g)/1e-9 for g in g_inhs])
isi = array([float(i)/1e-3 for i in isis])
scatter(g_inh[g_inh<4], isi[g_inh<4],color='k')
xlim([0,4])
tick_params(labelsize=14)
xlabel('synaptic conductance (nS)',fontsize=16)
ylabel('ISI (ms)',fontsize=16)
simpleaxis(ax)
fig.subplots_adjust(wspace=.5)
```
