<h2>Experiments with BS & PKJ ISIs<h2>

The goal of this experiment is to reciprically connect a network of BS and PKJ neurons and observe the the change in ISI statistics given some set of synaptic weight values

Discussion: The mean and variance of spontaneous currents for BS cells are the same.  The same holds for PKJ cells (but with different means and variances).  By connecting this network with BS->BS, BS->PKJ and PKJ->BS synapses, the mean firing rates of both BS and PKJ cells decrease and the ISI variances increase, as observed in Hausser and Clark 1997.  The means are heavily affected by synaptic weights and number of synapses each neuron receives.  So, choosing the weights and number of synapses carefully one can achieve the target mean firing rate.  The variances of ISIs are more difficult.  Interestingly, the variance of BS neurons increase substantially.  The coefficient of variation (CV) for BS cells reaches ~.34 in this set up (~.40 in Hausser and Clark 1997 control).  Interestly, the ISI CV for PKJs stays rather low ~.23 here vs .49 in H&C 1997.  Additionaly, we see an increased asymmetry relative to GABA block in these simulations. Another interesting observation is that the PKJ->BS synapses have a significant affect on the ISI variance in BS cells: ISI variance without PKJ->BS synapses falls to ~.26.

By adding PKJ -> PKJ synapses, the ISI variance of PKJs increases substantially.

```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../')
from neuron_models import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
defaultclock.dt = .5*ms
```

```python
T = 60000
N_BS = 16
N_PKJ = 16
BS = BasketCellGroup(N_BS)
PKJ = PurkinjeCellGroup(N_PKJ)
```

```python
# synaptic weights
w_bs_pkj = 3.
w_bs_bs = w_bs_pkj
w_pkj_bs = 3.
w_pkj_pkj = 3.

# Synapses
S_BS_PKJ = Synapses(BS,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w_bs_pkj')
S_BS_BS = Synapses(BS,BS,model='w:1',pre='g_inh+=BS.g_inh_*w_bs_bs')
S_PKJ_BS = Synapses(PKJ,BS,model='w:1',pre='g_inh+=BS.g_inh_*w_pkj_bs')
S_PKJ_PKJ = Synapses(PKJ,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w_pkj_pkj')

# Connections
S_BS_PKJ[:,:] = 'minimum(abs(i-j),abs(abs(i-j)-N_PKJ))<=2' 
S_BS_BS[:,:] = '(minimum(abs(i-j),abs(abs(i-j)-N_BS))<=2) & (i!=j)'
S_PKJ_BS.connect_one_to_one()
S_PKJ_PKJ[:,:] = '(minimum(abs(i-j),abs(abs(i-j)-N_PKJ))<=1) & (i!=j)'
```

```python
@network_operation(Clock(dt=1*ms))
def random_current():
    bs_i = .0635 + .045*randn(len(BS))
    bs_i[bs_i<0] = 0.0
    BS.I = bs_i * nA
    pkj_i = .08 + .075*randn(len(PKJ))
    pkj_i[pkj_i<0] = 0.
    PKJ.I = pkj_i * nA
```

```python
# Monitor
MS_BS = SpikeMonitor(BS)
MR_BS = PopulationRateMonitor(BS,bin=1*ms)
MISI_BS = ISIHistogramMonitor(BS,bins=arange(0,120,2)*ms)

MS_PKJ = SpikeMonitor(PKJ)
MR_PKJ = PopulationRateMonitor(PKJ,bin=1*ms)
MISI_PKJ = ISIHistogramMonitor(PKJ,bins=arange(0,120,2)*ms)

start = time.time()
run(T*msecond)
print time.time() - start
```

```python
def isi_mean_and_std(monitor):
    '''
    compute the mean and variance of interspike intervals
    of a group of neurons
    '''
    isi = []
    for n_ind, times in monitor.spiketimes.iteritems():
        isi += list(diff(times)*1000)
    return mean(isi), var(isi)**.5
```

```python
hist_plot(MISI_BS)
mew, std = isi_mean_and_std(MS_BS)
s = 'mean rate = %0.1f Hz\nmean ISI = %0.1f ms\nISI std = %0.1f ms\nCV = %0.2f' % (mean(MR_BS.rate),mew,std,std/mew)
annotate(s,(20,0),xytext=(80,MISI_BS.count.max()*.5),xycoords='data',fontsize=15)
title('BS ISI Histogram')
xlim([0,150])

hist_plot(MISI_PKJ)
mew, std = isi_mean_and_std(MS_PKJ)
s = 'mean rate = %0.1f Hz\nmean ISI = %0.1f ms\nISI std = %0.1f ms\nCV = %0.2f' % (mean(MR_PKJ.rate),mew,std,std/mew)
annotate(s,(20,0),xytext=(50,MISI_PKJ.count.max()*.7),xycoords='data',fontsize=15)
title('PKJ ISI Histogram')
xlim([0,120])
```
