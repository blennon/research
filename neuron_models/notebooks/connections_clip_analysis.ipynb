```python
def unpack_values(data):
    '''
    unpack the data and return an array of the values
    
    data = [(neuron_id, value)...]
    '''
    return array(zip(*data)[1])

def compute_statistics(values):
    '''
    compute statistics for quartiles and mean
    '''
    return [percentile(values,25), percentile(values,50), percentile(values,75), mean(values)]

def population_statistics(results, pruned_syn):
    '''
    compute the population statistics for MLIs and PKJs for the case where
    synapses of type 'pruned_syn' have been pruned
    '''    
    mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = {}, {}, {}, {}
    for r in results:
        syn, pct, mli_fr, mli_cv, pkj_fr, pkj_cv = r
        if syn == pruned_syn:
            mli_fr_stats[pct] = compute_statistics(unpack_values(mli_fr))
            mli_cv_stats[pct] = compute_statistics(unpack_values(mli_cv))
            pkj_fr_stats[pct] = compute_statistics(unpack_values(pkj_fr))
            pkj_cv_stats[pct] = compute_statistics(unpack_values(pkj_cv))
    return  mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats
```

```python
def build_traces(frs_dict, cvs_dict, neuron_type):
    '''
    build representations of data for plotting.
    
    frs_dict: dictionary containing population firing rate statistics, key is the percent of pruning
    cvs_dict: similar, for ISI CVs
    neuron_type: PKJ or MLI
    
    returns plotly objects for plotting
    '''
    # set neuron type specific formatting options
    if neuron_type == 'MLI':
        c1, c2 = '#8C2318', 'rgb(224, 102, 102)'
        marker_color = 'black'
    elif neuron_type == 'PKJ':
        c1, c2 = '#0B486B', 'rgb(68, 122, 219)'
        marker_color = 'rgb(246, 178, 107)'
    else:
        raise Exception('neuron_type must be either MLI or PKJ')
        
    # traces for Firing Rates
    pcts, vals = map(squeeze,zip(*sorted(frs_dict.items())))
    fr_quartiles = Scatter(x=list(pcts-.0075),y=list(vals[:,1]),
                        error_y=ErrorY(
                            type='data',
                            symmetric=False,
                            array=vals[:,2]-vals[:,1],
                            arrayminus=vals[:,1]-vals[:,0],color=c1,thickness=4,width=6),#447adb
                        marker=Marker(size=14,color=c1),line=Line(width=4))
    fr_means = Scatter(x=pcts-.0075,y=vals[:,3],mode='markers',
                    marker=Marker(symbol='cross',size=10,color=marker_color))
    
    # traces for ISI CVs
    pcts, vals = map(squeeze,zip(*sorted(cvs_dict.items())))
    cv_quartiles = Scatter(x=pcts+.0075,y=vals[:,1],yaxis='y2',
                        error_y=ErrorY(
                            type='data',
                            symmetric=False,
                            array=vals[:,2]-vals[:,1],
                            arrayminus=vals[:,1]-vals[:,0], color=c2,thickness=4,width=6),#FF851B
                        marker=Marker(size=14, color=c2),line=Line(width=4))
    cv_means = Scatter(x=pcts+.0075,y=vals[:,3],mode='markers',yaxis='y2',
                    marker=Marker(symbol='cross',size=10, color=marker_color))
    return fr_quartiles, fr_means, cv_quartiles, cv_means

def create_fig(fr_stats, cv_stats, neuron_type):
    '''
    creates the plotly figure
    '''
    fr_quartiles, fr_means, cv_quartiles, cv_means = build_traces(fr_stats, cv_stats, neuron_type)
    data = Data([fr_quartiles, fr_means, cv_quartiles, cv_means])
    
    # set neuron type specific formatting options
    if neuron_type == 'MLI':
        y1_range, y2_range = [5,30] ,[0.,.8]
        c1, c2 = '#8C2318', 'rgb(224, 102, 102)'
    elif neuron_type == 'PKJ':
        y1_range, y2_range = [5,30] ,[0.,.8]
        c1, c2 = '#0B486B', 'rgb(68, 122, 219)'
    else:
        raise Exception('neuron_type must be either MLI or PKJ')
        
    layout = Layout(
        xaxis=XAxis(title='Percent of synapses pruned',showgrid=False,autotick=False,dtick=.25,zeroline=False),
        yaxis=YAxis(
            title='Firing Rate (Hz)',
            showgrid=False,
            range=y1_range,
            titlefont=Font(color=c1),
            tickfont=Font(color=c1)
        ),
        yaxis2=YAxis(
            title='ISI CV',
            showgrid=False,
            side='right',
            overlaying='y',
            range=y2_range,
            titlefont=Font(color=c2),
            tickfont=Font(color=c2)
        ),
        font=Font(size=28, color='black'),
        autosize=False,
        width=900,
        height=500,
        boxmode='group',
        showlegend=False,
        margin=Margin(
            l=80,
            r=80,
            b=80,
            t=50,
            pad=0)
    )
    return PyFigure(data=data, layout=layout)
```

```python
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = population_statistics(results, 'MLI_MLI')
py.iplot(create_fig(mli_fr_stats,mli_cv_stats, 'MLI'), world_readable=False)
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = population_statistics(results, 'PKJ_MLI')
py.iplot(create_fig(mli_fr_stats,mli_cv_stats, 'MLI'), world_readable=False)
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = population_statistics(results, 'MLI_MLI')
py.iplot(create_fig(pkj_fr_stats,pkj_cv_stats, 'PKJ'), world_readable=False)
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = population_statistics(results, 'PKJ_MLI')
py.iplot(create_fig(pkj_fr_stats,pkj_cv_stats, 'PKJ'), world_readable=False)
```

<h2>Hypothesis testing</h2>

```python
from scipy.stats import shapiro, mannwhitneyu
```

```python
def neuron_statistics(results, pruned_syn):    
    mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = {}, {}, {}, {}
    for r in results:
        syn, pct, mli_fr, mli_cv, pkj_fr, pkj_cv = r
        if syn == pruned_syn:
            mli_fr_stats[pct] = unpack_values(mli_fr)
            mli_cv_stats[pct] = unpack_values(mli_cv)
            pkj_fr_stats[pct] = unpack_values(pkj_fr)
            pkj_cv_stats[pct] = unpack_values(pkj_cv)
    return  mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats
```

<h3>Statistical tests for differences in population statistics between no pruning and completely pruned MLI-MLI synapses</h3>

We can reject the null hypothesis that the data come from the same distribution. This suggests MLI-MLI synapses have a significant effect on the baseline state of the model network.

```python
# MLI_MLI prune
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = neuron_statistics(results,'MLI_MLI')

# Normality tests, p<.05 reject null hypothesis (normal)
print 'Shapiro-Wilk test for Normality, no pruning, W: %s, p: %s' % shapiro(mli_fr_stats[0.0])
print 'Shapiro-Wilk test for Normality, complete pruning, W: %s, p: %s' % shapiro(mli_fr_stats[1.0])
print 'Shapiro-Wilk test for Normality, no pruning, W: %s, p: %s' % shapiro(mli_cv_stats[0.0])
print 'Shapiro-Wilk test for Normality,  complete pruning, W: %s, p: %s' % shapiro(mli_cv_stats[1.0])

# Test differences in population, p<.05 reject null hypothesis (same distribution)
print 'Mann-Whitney U test for MLI FRs, U: %s, p: %s' % mannwhitneyu(mli_fr_stats[0.0], mli_fr_stats[1.0])
print 'Mann-Whitney U test for MLI ISI CVs, U: %s, p: %s' % mannwhitneyu(mli_cv_stats[0.0], mli_cv_stats[1.0])
print 'Mann-Whitney U test for PKJ FRs, U: %s, p: %s' % mannwhitneyu(pkj_fr_stats[0.0], pkj_fr_stats[1.0])
print 'Mann-Whitney U test for PKJ ISI CVs, U: %s, p: %s' % mannwhitneyu(pkj_cv_stats[0.0], pkj_cv_stats[1.0])
```

<h3>Statistical tests for differences in population statistics between no pruning and completely pruned PKJ-MLI synapses</h3>
We cannot reject the null hypothesis that the data come from the same distribution. Thus, the PKJ-MLI synapses have a negligle effect on the default state of the model network.

```python
# PKJ_MLI prune
mli_fr_stats, mli_cv_stats, pkj_fr_stats, pkj_cv_stats = neuron_statistics(results,'PKJ_MLI')

# Normality tests, p<.05 reject null hypothesis (normal)
print 'Shapiro-Wilk test for Normality, no pruning, W: %s, p: %s' % shapiro(mli_fr_stats[0.0])
print 'Shapiro-Wilk test for Normality, complete pruning, W: %s, p: %s' % shapiro(mli_fr_stats[1.0])
print 'Shapiro-Wilk test for Normality, no pruning, W: %s, p: %s' % shapiro(mli_cv_stats[0.0])
print 'Shapiro-Wilk test for Normality,  complete pruning, W: %s, p: %s' % shapiro(mli_cv_stats[1.0])

# Test differences in population, p<.05 reject null hypothesis (same distribution)
print 'Mann-Whitney U test for MLI FRs, U: %s, p: %s' % mannwhitneyu(mli_fr_stats[0.0], mli_fr_stats[1.0])
print 'Mann-Whitney U test for MLI ISI CVs, U: %s, p: %s' % mannwhitneyu(mli_cv_stats[0.0], mli_cv_stats[1.0])
print 'Mann-Whitney U test for PKJ FRs, U: %s, p: %s' % mannwhitneyu(pkj_fr_stats[0.0], pkj_fr_stats[1.0])
print 'Mann-Whitney U test for PKJ ISI CVs, U: %s, p: %s' % mannwhitneyu(pkj_cv_stats[0.0], pkj_cv_stats[1.0])
```
