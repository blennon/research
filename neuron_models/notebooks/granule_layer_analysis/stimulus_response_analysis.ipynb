<h1>Stimulus response analysis</h1>

The goal of these experiments and analysis is to see how the Granule cell layer responses changes to different stimuli.
E.g. we would like to test:

* frequency
* amplitude
* phase
* wave form

Presently, frequency and wave from have been tested (see below).

This analysis uses the cross-correlation measure described in Yamazaki and Nagao 2012.

```python
%load_ext autoreload
%autoreload 2
import cPickle
import sys
sys.path.append('../')
from neuron_models import *
```

```python
data_dir = '/home/bill/research/data/neuron_models/granule_layer/'
N_GO = 32**2
N_GR = N_GO * 10**2
T = 2000
```

<h2> Same net, same stimulus twice </h2>

```python
with open(data_dir+'stimuli_par/2013-04-17T12:01:24.982327/sinusoid_one_period1_spikes.pickle') as inf:
    GR_spikes1 = cPickle.load(inf)
with open(data_dir+'stimuli_par/2013-04-17T12:01:24.982327/sinusoid_one_period2_spikes.pickle') as inf:
    GR_spikes2 = cPickle.load(inf)
```

```python
GR1_GR1_sim = population_spike_correlation(GR_spikes1,None,N_GR,N_GO,T)
GR1_GR2_sim = population_spike_correlation(GR_spikes1,GR_spikes2,N_GR,N_GO,T)
```

```python
figure(1,figsize=(10,8))
l1,l2 = plot(arange(-T/2,T/2-1),GR1_GR1_sim,'r',arange(-T/2,T/2-1),GR1_GR2_sim,'b')
legend([l1,l2],['Auto-correlation','Cross-correlation GR1-GR2'])
title('Stimulus response of the same network to the same stimulus',fontsize=20)
```

When the same net is repeatedly presented the same stimulus, the GR response is very similar.

<h2>Different net (weights), same stimulus</h2>

```python
GR1_GR3_sim = population_spike_correlation(GR_spikes1,GR_spikes3,N_GR,N_GO,T)
```

The GR responses of two different networks presented with the same stimuli has very low similarity.

<h2>Same net, different stimuli</h2>

```python
stimuli_names = ['triangle_one_period','triangle_two_periods','square_one_period_in_phase',
                 'square_two_periods_in_phase','sinusoid_one_period','sinusoid_two_periods']                 
stimuli_responses = {}
for sn in stimuli_names:
    with open(data_dir+'/stimuli_par/2013-04-16T13:31:30.462202/'+sn+'_spikes.pickle') as inf:
        stimuli_responses[sn] = cPickle.load(inf)
        inf.close()
```

<h3>GR population responses</h3>

```python
fig2 = figure(2,figsize(25,8))
i = 1
for sn in sorted(stimuli_responses.keys()):
    ax = fig2.add_subplot(2,3,i)
    plot_raster_firingrate_overlay(stimuli_responses[sn],T,range(0,N_GR,N_GR/N_GO),ax)
    title(sn.replace('_',' '),fontsize=18)
    i += 1
```

<h3>Correlation measures</h3>

```python
stimuli_correlations = {}
for s1 in stimuli_responses:
    for s2 in stimuli_responses:
        if (s1,s2) in stimuli_correlations or (s2,s1) in stimuli_correlations:
            continue
        corr = population_spike_correlation(stimuli_responses[s1],stimuli_responses[s2],N_GR,N_GO,T)
        stimuli_correlations[(s1,s2)] = corr
```

```python
i = 1
fig3 = figure(3,figsize(30,15))
for s1 in stimuli_responses:
    subplot(2,3,i)
    lines = []
    for s2 in stimuli_responses:
        if (s1,s2) in stimuli_correlations:
            sim = stimuli_correlations[(s1,s2)]
        else:
            sim = stimuli_correlations[(s2,s1)]
        l, = plot(arange(-T/2,T/2-1),sim)
        ylim([.5,1.])
        lines.append(l)
    legend(lines,stimuli_responses.keys())
    title(s1.replace('_',' '),fontsize=18)
    i += 1
```

As we can see from the plots, the measure of simility of GR response to sine waves and triangle waves is pretty close; as close in similarity as the same sine wave presented twice to the same network (see plots above).  As we can see from the population firing rates, the responses are a bit different.  One issue is that the correlation measure blurs or loses some information.  I calculated correlations again with smaller tau but qualitatively the issue is the same.

Perhaps slightly randomizing the synaptic weights or neuron parameters could result in further differentiation in GR responses.
