<h2>Experiments with MLI
    & PKJ ISIs<h2>

The goal of this experiment is to add additional realism to the anatomical connectivity of the network.  E.g. random weights, increased convergence/divergence ratios, etc.

```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../../')
from neuron_models import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
defaultclock.dt = .25*ms
from statsmodels.tsa.stattools import acf
from pandas import *
```

```python
T = 4*second
N_GO = 32**2
N_GR = N_GO * 5**2
N_MF = N_GR
N_MLI = 160
N_PKJ = 16
ltd_window = 50*ms
MLI = MLIGroup(N_MLI)
PKJ = PurkinjeCellGroup(N_PKJ)
GO = GolgiCellGroup(N_GO)
GR = GranuleCellGroup(N_GR,max_delay=ltd_window)
MF = PoissonGroup(N_MF, lambda t: stimulus(t,15*Hz,15*Hz,pi*Hz,pi,lambda x: 1+cos(x)))
MF_IO = PoissonGroup(1, lambda t: stimulus(t,0*Hz,3*Hz,pi*Hz,pi,lambda x: 1+cos(x)))
IO = InferiorOliveGroup(1)
```

```python
syn_dir = '/home/bill/research/data/neuron_models/okr/synapses_11.5.13/'
create_connections = False
save_connections = False

# synaptic weights
w_mli_pkj = 1.#.9
w_mli_mli = 1.#0.3
w_pkj_mli = 1.#67
w_pkj_pkj = .65#.31
w_pf_mli = 1.
w_pf_pkj = .003
w_gr_go = .2/(49*(len(GR)/len(GO)))
w_go_gr = 20.
w_mf_gr = 4.
w_mf_io = 1.0

# LTP, LTD coefficients
c_ltp = .0005
c_ltd = .995
c_pf_mli_ltd = .0005
c_pf_mli_ltp = .005

# Synapses
S_GO_GR = Synapses(GO,GR,model='w:1',pre='g_inh1+=GR.g_inh_*w_go_gr; g_inh2+=GR.g_inh_*w_go_gr')
S_GR_GO = Synapses(GR,GO,model='w:1',pre='g_ampa+=GO.g_ampa_*w_gr_go;g_nmda1+=GO.g_ampa_*w_gr_go;g_nmda2+=GO.g_ampa_*w_gr_go')
S_MF_GR = Synapses(MF,GR,model='w:1',pre='g_ampa+=GR.g_ampa_*w_mf_gr;g_nmda+=GR.g_ampa_*w_mf_gr')
S_MLI_PKJ = Synapses(MLI,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w')
S_MLI_MLI = Synapses(MLI,MLI,model='w:1',pre='g_inh+=MLI.g_inh_*w')
S_PKJ_MLI = Synapses(PKJ,MLI,model='w:1',pre='g_inh+=MLI.g_inh_*w')
S_PKJ_PKJ = Synapses(PKJ,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w')
S_PF_PKJ = Synapses(GR,PKJ,model='''w:1''',pre='''g_ampa+=PKJ.g_ampa_*w; w+=c_ltp*(w_pf_pkj-w)''')
S_PF_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w; w-=c_pf_mli_ltd*w')
S_MF_IO_IO = Synapses(MF_IO,IO,model='w:1',pre='g_ampa+=IO.g_ampa_*w_mf_io')
S_CF_PKJ = Synapses(IO,PKJ,model='w:1')
S_CF_MLI = Synapses(IO,MLI,model='w:1')

# Connections
if create_connections:
    S_MF_GR.connect_one_to_one()
    pre, post = gr_to_go_connections(N_GO,N_GR)
    S_GR_GO.create_synapses(pre,post)
    pre, post = go_to_gr_connections(N_GO,N_GR)
    S_GO_GR.create_synapses(pre,post)
    pre, post = gr_to_pkj_connections(N_GO,N_GR,N_PKJ)
    connect_mli_pkj(S_MLI_PKJ,pkj_dist=8,syn_prob=.25)
    connect_mli_mli(S_MLI_MLI,dist=80,syn_prob=.05)
    S_PKJ_MLI[:,:] = '(j/(N_MLI/N_PKJ) == i) & (rand()<.5)'
    S_PKJ_PKJ[:,:] = '(minimum(abs(i-j),abs(abs(i-j)-N_PKJ))<=5) & (i!=j) & (rand()<.25)'
    S_MLI_PKJ.w[:,:] = 'rand()*w_mli_pkj'
    S_MLI_MLI.w[:,:] = 'rand()*w_mli_mli'
    S_PKJ_MLI.w[:,:] = 'rand()*w_pkj_mli'
    S_PKJ_PKJ.w[:,:] = 'rand()*w_pkj_pkj'
    S_MF_IO_IO.connect_one_to_one()
    S_PF_PKJ.create_synapses(*gr_to_pkj_connections(N_GO,N_GR,N_PKJ))
    S_PF_PKJ.w[:,:] = .003
    S_PF_MLI.create_synapses(*gr_to_mli_connections(N_GO,N_GR,N_PKJ,p=.007))
    S_PF_MLI.w[:,:] = .01
    S_CF_PKJ[:,:] = True
    S_CF_MLI[:,:] = True
else:
    load_synapses(S_GO_GR, 'S_GO_GR', syn_dir)
    load_synapses(S_GR_GO, 'S_GR_GO', syn_dir)
    load_synapses(S_MF_GR, 'S_MF_GR', syn_dir)
    load_synapses(S_MLI_PKJ, 'S_MLI_PKJ', syn_dir)
    load_synapses(S_MLI_MLI, 'S_MLI_MLI', syn_dir)
    load_synapses(S_PKJ_MLI, 'S_PKJ_MLI', syn_dir)
    load_synapses(S_PKJ_PKJ, 'S_PKJ_PKJ', syn_dir)
    load_synapses(S_PF_PKJ, 'S_PF_PKJ', syn_dir)
    load_synapses(S_PF_MLI, 'S_PF_MLI', syn_dir)
    load_synapses(S_MF_IO_IO, 'S_MF_IO_IO', syn_dir)
    load_synapses(S_CF_PKJ, 'S_CF_PKJ', syn_dir)
    load_synapses(S_CF_MLI, 'S_CF_MLI', syn_dir)

#S_MLI_PKJ.w[:,:] *= 0
#S_PKJ_PKJ.w[:,:] *= 0
#S_MLI_MLI.w[:,:] *= 0
#S_PKJ_MLI.w[:,:] *= 0

print 'Convergences'
print 'MLI->PKJ convergence: ', len(S_MLI_PKJ)/float(N_PKJ)
print 'MLI->MLI convergence: ', len(S_MLI_MLI)/float(N_MLI)
print 'PKJ->MLI convergence: ', len(S_PKJ_MLI)/float(N_MLI)
print 'PKJ->PKJ convergence: ', len(S_PKJ_PKJ)/float(N_PKJ)
print 'PF->MLI convergence: ', len(S_PF_MLI)/float(N_MLI)
print 'PF->PKJ convergence: ', len(S_PF_PKJ)/float(N_PKJ)
print '\nDivergences'
print 'MLI->PKJ divergence: ', len(S_MLI_PKJ)/float(N_MLI)
print 'MLI->MLI divergence: ', len(S_MLI_MLI)/float(N_MLI)
print 'PKJ->MLI divergence: ', len(S_PKJ_MLI)/float(N_PKJ)
print 'PKJ->PKJ divergence: ', len(S_PKJ_PKJ)/float(N_PKJ)
print 'PF->MLI divergence: ', len(S_PF_MLI)/float(N_GR)
print 'PF->PKJ divergence: ', len(S_PF_PKJ)/float(N_GR)

PF_PKJ_LTD = PF_PKJ_LTD(IO,GR,S_PF_PKJ,S_CF_PKJ,window=ltd_window)
PF_MLI_LTP = PF_MLI_LTP(IO,GR,S_PF_MLI,S_CF_MLI,c_pf_mli_ltp,w_pf_mli,window=ltd_window)
```

```python
def separate_spikes_across_trials(spiketimes, period, num_periods):
    '''
    Takes a dictionary of arrays, each containing spiketimes for a particular neuron
    across several trials and splits the array into spike times belonging to each
    trial period.

    spiketimes: dict {neuron_index:array_of_spiketimes,...}
    period: the time duration of each trial, in units of seconds
    num_periods: number of trials/periods represented in the spike times

    Returns a dictionary of arrays with additional keys, each belonging to a neuron
    for a particular trial
    '''
    new_spiketimes = {}
    for ind, times in spiketimes.iteritems():
        for i in range(num_periods):
            new_spiketimes[ind+len(spiketimes)*i] = times[(times>period*i) & (times<period*(i+1))] - period*i
    return new_spiketimes

fig = figure(figsize=(12,6))
ax = fig.add_subplot(111)
new_spiketimes = separate_spikes_across_trials(MS_PKJ.spiketimes, 4*second, 4)
plot_raster_firingrate_overlay(new_spiketimes, 4000, arange(0,N_PKJ), ax)
xlabel('Time (ms)',fontsize=18)
title('PKJ Raster', fontsize=20)
```
