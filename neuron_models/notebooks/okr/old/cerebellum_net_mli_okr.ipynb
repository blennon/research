```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../')
from neuron_models import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
defaultclock.dt = .5*ms
```

```python
T = 1*2000
N_GO = 32**2
N_GR = N_GO * 5**2
N_MF = N_GR
N_BS = 16
N_PKJ = 16
N_VN = 1
N_MF_PER_VN = 4
ltd_window = 50 * ms
GO = GolgiCellGroup(N_GO)
GR = GranuleCellGroup(N_GR,max_delay=ltd_window)
MF = PoissonGroup(N_MF, lambda t: stimulus(t,15*Hz,30*Hz,pi*Hz,pi,lambda x: .5*(1+cos(x))))
MF_IO = PoissonGroup(1, lambda t: stimulus(t,0*Hz,3*Hz,pi*Hz,pi,lambda x: .5*(1+cos(x))))
MF_VN = PoissonGroup(N_VN, lambda t: stimulus(t,0*Hz,30*Hz,pi*Hz,pi,lambda x: N_MF_PER_VN*.5*(1+cos(x))))
BS = BasketCellGroup(N_BS)
PKJ = PurkinjeCellGroup(N_PKJ)
IO = InferiorOliveGroup(1)
VN = VestibularNucleusGroup(N_VN)
```

```python
def w_eq_add(w_max, c_ltp, c_ltd, alpha, cf_rate):
    return w_max/(1+(c_ltd/c_ltp)*alpha*cf_rate)
def w_eq_mult(w_max, c_ltp, c_ltd, alpha, cf_rate, pf_rate):
    a = pf_rate*c_ltp/cf_rate
    return w_max/(1+(1-c_ltd**(alpha*pf_rate))/a)
def w_eq_add_ltp(w_max, c_ltp, c_ltd, cf_rate, alpha):
    a = c_ltd/(alpha*cf_rate*c_ltp)
    return w_max/(1+a)
```

```python
# synaptic weights
w_gr_go = .2/(49*(len(GR)/len(GO)))
w_go_gr = 20.
w_mf_gr = 4.
w_pf_bs = .003
w_pf_pkj = w_eq_mult(.003,.0001,.995,50*msecond,1*Hz,5*Hz)
w_bs_pkj = N_BS/3
w_cf_pkj = 0.
w_cf_bs = 0.
w_mf_io = 1.0
w_vn_io = 4.0
w_pkj_vn = .12/N_PKJ
w_mf_vn = .2/N_MF_PER_VN

# LTP,LTD coeffs
c_ltp = .0001
c_ltd = .995
c_pf_mli_ltd = .0001
c_pf_mli_ltp = .005

# Synapses
S_GO_GR = Synapses(GO,GR,model='w:1',pre='g_inh1+=GR.g_inh_*w_go_gr; g_inh2+=GR.g_inh_*w_go_gr')
S_GR_GO = Synapses(GR,GO,model='w:1',pre='g_ampa+=GO.g_ampa_*w_gr_go;g_nmda1+=GO.g_ampa_*w_gr_go;g_nmda2+=GO.g_ampa_*w_gr_go')
S_MF_GR = Synapses(MF,GR,model='w:1',pre='g_ampa+=GR.g_ampa_*w_mf_gr;g_nmda+=GR.g_ampa_*w_mf_gr')
S_PF_PKJ = Synapses(GR,PKJ,model='''w:1''',pre='''g_ampa+=PKJ.g_ampa_*w_pf_pkj''')
S_PF_BS = Synapses(GR,BS,model='w:1',pre='g_ampa+=BS.g_ampa_*w; w-=c_pf_mli_ltd*w')
S_BS_PKJ = Synapses(BS,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w_bs_pkj')
S_CF_PKJ = Synapses(IO,PKJ,model='w:1',pre='g_ampa+=PKJ.g_ampa_*w_cf_pkj')
S_CF_BS = Synapses(IO,BS,model='w:1',pre='g_ampa+=BS.g_ampa_*w_cf_bs')
S_MF_IO_IO = Synapses(MF_IO,IO,model='w:1',pre='g_ampa+=IO.g_ampa_*w_mf_io')
S_VN_IO = Synapses(VN,IO,model='w:1',pre='g_gaba+=IO.g_gaba_*w_vn_io')
S_PKJ_VN = Synapses(PKJ,VN,model='w:1',pre='g_gaba+=VN.g_gaba_*w_pkj_vn')
S_MF_VN = Synapses(MF_VN,VN,model='w:1',pre='g_ampa+=VN.g_ampa_*w_mf_vn; g_nmda+=VN.g_ampa_*w_mf_vn')

# Connections
S_MF_GR.connect_one_to_one()
pre, post = gr_to_go_connections(N_GO,N_GR)
S_GR_GO.create_synapses(pre,post)
pre, post = go_to_gr_connections(N_GO,N_GR)
S_GO_GR.create_synapses(pre,post)
pre, post = gr_to_pkj_connections(N_GO,N_GR,N_PKJ)
S_PF_BS.create_synapses(pre,post)
S_PF_BS.w = w_eq_add_ltp(.003,.005,.0001,3*Hz,50*msecond)
S_PF_PKJ.create_synapses(pre,post)
S_BS_PKJ[:,:]='abs(i-j)%(N_PKJ-1) < 2' # nearest three BS -> PKJ
S_CF_PKJ[:,:] = True
S_CF_BS[:,:] = True
S_MF_IO_IO.connect_one_to_one()
S_VN_IO.connect_one_to_one()
S_PKJ_VN[:,:] = True
S_MF_VN.connect_one_to_one()

# LTD
#PF_PKJ_LTD = LTD(IO,GR,S_PF_PKJ,S_CF_PKJ,window=ltd_window)
# PF_MLI_LTP
LTP = PF_MLI_LTP(IO,GR,S_PF_BS,S_CF_BS,c_pf_mli_ltp,w_pf_bs,window=ltd_window)
```

```python
MS_MF_IO = SpikeMonitor(MF_IO)
MS_MF_VN = SpikeMonitor(MF_VN)
MS_VN = SpikeMonitor(VN)
MS_IO = SpikeMonitor(IO)
MR_GR = PopulationRateMonitor(GR,bin=1*ms)
MR_GO = PopulationRateMonitor(GO,bin=1*ms)
MR_BS = PopulationRateMonitor(BS,bin=1*ms)
MR_PKJ = PopulationRateMonitor(PKJ,bin=1*ms)
weights = random_integers(len(S_PF_BS),size=100)
MW_PF_BS = StateMonitor(S_PF_BS,'w',record=weights)
start = time.time()
run(T*msecond)
print time.time() - start
```
