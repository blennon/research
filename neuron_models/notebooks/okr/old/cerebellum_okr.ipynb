```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../../')
from neuron_models import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
defaultclock.dt = .5*ms
```

```python
def equilibrium_weights(pf_rate, cf_rate, c_ltp, c_ltd, w_max, ltd_window):
    r = pf_rate/cf_rate
    return (r*c_ltp*w_max)/(1-c_ltd**(ltd_window*pf_rate)+r*c_ltp)
```

```python
T = 1*2000
N_GO = 32**2
N_GR = N_GO * 5**2
N_MF = N_GR
N_BS = 16
N_PKJ = 16
N_VN = 1
N_MF_PER_VN = 4
ltd_window = 50 * ms
GO = GolgiCellGroup(N_GO)
GR = GranuleCellGroup(N_GR,max_delay=ltd_window)
MF = PoissonGroup(N_MF, lambda t: stimulus(t,15*Hz,15*Hz,pi*Hz,pi,lambda x: 1+cos(x)))
MF_IO = PoissonGroup(1, lambda t: stimulus(t,0*Hz,3*Hz,pi*Hz,pi,lambda x: 1+cos(x)))
MF_VN = PoissonGroup(N_VN, lambda t: stimulus(t,0*Hz,15*Hz,pi*Hz,pi,lambda x: N_MF_PER_VN*(1+cos(x))))
BS = BasketCellGroup(N_BS)
PKJ = PurkinjeCellGroup(N_PKJ)
IO = InferiorOliveGroup(1)
VN = VestibularNucleusGroup(N_VN)
```

```python
# synaptic weights
w_gr_go = .2/(49*(len(GR)/len(GO)))
w_go_gr = 20.
w_mf_gr = 4.
w_pf_bs = .003#300./N_GR
w_pf_pkj = .003
w_bs_pkj = N_BS/3
w_cf_pkj = 0.
w_mf_io = 1.0
w_vn_io = 4.0
w_pkj_vn = .12/N_PKJ
w_mf_vn = .2/N_MF_PER_VN

# LTP,LTD coeffs
c_ltp = .0001
c_ltd = .995

# Synapses
S_GO_GR = Synapses(GO,GR,model='w:1',pre='g_inh1+=GR.g_inh_*w_go_gr; g_inh2+=GR.g_inh_*w_go_gr')
S_GR_GO = Synapses(GR,GO,model='w:1',pre='g_ampa+=GO.g_ampa_*w_gr_go;g_nmda1+=GO.g_ampa_*w_gr_go;g_nmda2+=GO.g_ampa_*w_gr_go')
S_MF_GR = Synapses(MF,GR,model='w:1',pre='g_ampa+=GR.g_ampa_*w_mf_gr;g_nmda+=GR.g_ampa_*w_mf_gr')
S_PF_PKJ = Synapses(GR,PKJ,model='''w:1''',pre='''g_ampa+=PKJ.g_ampa_*w; w+=c_ltp*(w_pf_pkj-w)''')
S_PF_BS = Synapses(GR,BS,model='w:1',pre='g_ampa+=BS.g_ampa_*w_pf_bs')
S_BS_PKJ = Synapses(BS,PKJ,model='w:1',pre='g_inh+=PKJ.g_inh_*w_bs_pkj')
S_CF_PKJ = Synapses(IO,PKJ,model='w:1',pre='g_ampa+=PKJ.g_ampa_*w_cf_pkj')
S_MF_IO_IO = Synapses(MF_IO,IO,model='w:1',pre='g_ampa+=IO.g_ampa_*w_mf_io')
S_VN_IO = Synapses(VN,IO,model='w:1',pre='g_inh+=IO.g_inh_*w_vn_io')
S_PKJ_VN = Synapses(PKJ,VN,model='w:1',pre='g_inh+=VN.g_inh_*w_pkj_vn')
S_MF_VN = Synapses(MF_VN,VN,model='w:1',pre='g_ampa+=VN.g_ampa_*w_mf_vn; g_nmda+=VN.g_ampa_*w_mf_vn')

# Connections
S_MF_GR.connect_one_to_one()
pre, post = gr_to_go_connections(N_GO,N_GR)
S_GR_GO.create_synapses(pre,post)
pre, post = go_to_gr_connections(N_GO,N_GR)
S_GO_GR.create_synapses(pre,post)
pre, post = gr_to_pkj_connections(N_GO,N_GR,N_PKJ)
S_PF_BS.create_synapses(pre,post)
S_PF_PKJ.create_synapses(pre,post)
S_PF_PKJ.w = w_pf_pkj
S_BS_PKJ[:,:]='abs(i-j)%(N_PKJ-1) < 2' # nearest three BS -> PKJ
S_CF_PKJ[:,:] = True
S_MF_IO_IO.connect_one_to_one()
S_VN_IO.connect_one_to_one()
S_PKJ_VN[:,:] = True
S_MF_VN.connect_one_to_one()

# LTD
PF_PKJ_LTD = PF_PKJ_LTD(IO,GR,S_PF_PKJ,S_CF_PKJ,window=ltd_window)

# Spontaneous current
@network_operation
def spontaneous_current():
    PKJ.I = .25 * nA
```

```python
#MS_MF = SpikeMonitor(MF)
MS_MF_IO = SpikeMonitor(MF_IO)
MS_MF_VN = SpikeMonitor(MF_VN)
MS_VN = SpikeMonitor(VN)
MS_IO = SpikeMonitor(IO)
#MS_GR = SpikeMonitor(GR)
#MS_GO = SpikeMonitor(GO)
#MS_BS = SpikeMonitor(BS)
#MS_PKJ = SpikeMonitor(PKJ)
MR_GR = PopulationRateMonitor(GR,bin=1*ms)
MR_GO = PopulationRateMonitor(GO,bin=1*ms)
MR_BS = PopulationRateMonitor(BS,bin=1*ms)
MR_PKJ = PopulationRateMonitor(PKJ,bin=1*ms)
weights = random_integers(len(S_PF_PKJ),size=100)
MW_PF_PKJ = StateMonitor(S_PF_PKJ,'w',record=weights)
start = time.time()
run(T*msecond)
print time.time() - start
```
