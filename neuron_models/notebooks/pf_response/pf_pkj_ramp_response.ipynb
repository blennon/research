The goal of this notebook is to test the ficticious PF burst input to a PKJ without spontaneous depolarizing current.  The PF input will consist of a burst of 6 spikes at 333 Hz.  We will determine the synaptic weight such that the PKJ depolarizes on the 6th spike.  w = 1

```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../../')
from neuron_models import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
defaultclock.dt = .25*ms
```

```python
def ramp_current(t,t0,t1,I_max):
    if t<t0: return 0.
    elif t>t1: return I_max
    else: return I_max/(t1-t0)*(t-t0)
```

```python
T = 500*ms
MLI = MLIGroup(1)
MLI.V = MLI.El
PKJ = PurkinjeCellGroup(1)
PKJ.V = PKJ.El

@network_operation(Clock(dt=defaultclock.dt))
def random_current():
    PKJ.I = gamma(0.430303,0.195962,size=len(PKJ)) * nA + ramp_current(defaultclock.t,50*ms,300*ms,.15*nA)
    MLI.I = ramp_current(defaultclock.t, 50*ms,300*ms,.07*nA) + gamma(3.966333,0.006653,size=len(MLI)) * nA

MS_PKJ = SpikeMonitor(PKJ)
MV_PKJ = StateMonitor(PKJ, 'V', record=0)
MS_MLI = SpikeMonitor(MLI)
MV_MLI = StateMonitor(MLI, 'V', record=0)
run(T)
```

```python
MV_MLI.plot()
tick_params(labelsize=16)
xlabel('time (s)',fontsize=16)
ylabel('Membrane Potential (V)',fontsize=16)

spiketimes = MS_MLI.spiketimes[0]
figure()
firing_rate(spiketimes[spiketimes>300*ms])
figure()
plot(1./diff(spiketimes))
```
