```python
from rate_model import *
```

```python
class Network(Network):
    # define neuron groups
    PF = NeuronGroup(2)
    MLI = ShuntingNeuronGroup(1, A=1.5, resting_state=array([0.1]))

    # connect neurons
    C_PF_MLI = PF_MLI_Connection(PF,MLI,W=array([[.1,.1]]))
    MLI.connect(C_PF_MLI)

    # monitor neuron & weight states
    PF_monitor = StateMonitor(PF)
    MLI_monitor = StateMonitor(MLI)
    W_monitor = StateMonitor(C_PF_MLI)
    
    def update(self):
        '''
        Update the state of the network
        '''
        self.MLI.update()
        self.C_PF_MLI.update(beta=.1)
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.W_monitor.record()
        
    def run(self):
        for i in range(350):
            if 0<=i<50:
                self.PF.set_state(array([0.,0.]))
            elif 50<=i<100:
                self.PF.set_state(array([1.,0.]))
            elif 100<=i<150:
                self.PF.set_state(array([1.,1.]))
            elif 150<=i<200:
                self.PF.set_state(array([0.,1.]))
            elif 200<=i<250:
                self.PF.set_state(array([0.,0.]))
            elif 250<=i<300:
                self.PF.set_state(array([1.,0.]))
            elif 300<=i<350:
                self.PF.set_state(array([0.,0.]))
            self.update()
```

```python
net = Network()
net.run()
```

```python
def set_axes():
    xlabel('time (ms)',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)
    
figure(figsize=(8,4))
net.MLI_monitor.plot(color='r', lw=2., label='MLI Firing rate')
legend(loc='upper right', fontsize=14)
set_axes()
title('MLI Firing rate', fontsize=16)

def plot_pf(ind, c='b'): 
    net.PF_monitor.plot(ind, color=c, lw=2., label='PF%s firing rate'%ind)
    net.W_monitor.plot(ind, color=c, lw=2., ls='--', label='PF%s-MLI weight'%ind)
    legend(loc='upper center', fontsize=14)
    set_axes()
    ylim([0,2])
    title('PF%s Firing rate'%ind, fontsize=16)
    
figure(figsize=(8,4))
plot_pf(0)
figure(figsize=(8,4))
plot_pf(1,'g')
```
