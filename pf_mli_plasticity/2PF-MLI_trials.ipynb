<h1>PF-MLI Learning Simulations</h1>

This IPython notebook simulates a simple network consisting of one molecular layer interneuron (MLI), two parallel fibers (PFs) and one climbing fiber (CF). In particular, we study activity dependent changes in PF-MLI synaptic weights.

The model consists of two PFs: $PF_{1}$ and $PF_{2}$ that make synapses onto the MLI, i.e. $w_{1}$ and $w_{2}$. The MLI also receives input from a single CF with synaptic weight $w_{CF}$.

$$\tau_{}\frac{d\mathbf{\mathrm{M}\mathrm{L}\mathrm{I}}\mathcal{}\ }{dt}=-A(\mathbf{\mathrm{M}\mathrm{L}\mathrm{I}}-MLI_{rest})+(C-D\mathbf{\mathrm{M}\mathrm{L}\mathrm{I}})\sum_{i}w_{i}\mathrm{P}\mathrm{F}_{i}(t)$$

where, $\mathrm{M}\mathrm{L}\mathrm{I}$ is the firing rate of the neuron, $\tau=1$ is a time constant, $A$ governs the passive decay of the firing rate, $MLI_{rest}$ is the baseline firing rate of the neuron, $\mathrm{P}\mathrm{F}_{i}(t)\in[0,1]$ is the firing rate of the $i^{th}$ PF input, and $w_{i}\in[0,1]$ is the synaptic weight corresponding to $\mathrm{P}\mathrm{F}_{i}(t)$.

The learning rule for the PF-MLI synapses is governed by:

$$\frac{dw_{i} }{dt} = \begin{cases} \beta\mathrm{P}\mathrm{F}_{i}(t)[\mathbf{\mathrm{M}\mathrm{L}\mathrm{I}}_{j}(t)-\alpha w_{i}], & \mbox{if } w_{i}>0 \\ 0, & \mbox{if } w_{i}=0\mbox{ and } \mathrm{P}\mathrm{F_{i}(t)}=0 \\ \delta, & \mbox{if }w_{i}=0 \mbox{ ,} \mathrm{P}\mathrm{F_{i}(t)}>0,\mathrm{C}\mathrm{F(t)}>0 \end{cases}$$

where $\beta$ is the learning rate and $\delta=0.1$ is the initial weight value for newly active synapses.

We simulate the changes in $w_{1}$ and $w_{2}$ in under different conditions and show that the model is capable of reproducing in vitro experimental results.

<h3>Define Network setup</h3>

```python
class Simulation(object):
    '''
    Simulates the activity of a Network object
    '''
    
    def __init__(self, net, dt=.1):
        '''
        net: Network object
        dt: simulation time step
        '''
        self.dt = dt
        self.net = net
    
    def run(self):
        for t in range(250):
            # update state every dt
            for i in range(int(1./self.dt)):
                self.set_input_states(t)
                self.net.update()
            # update monitors every ms
            self.net.record()
    
    def set_input_states(self, i):
        PF1_max = self.net.pf1
        PF2_max = self.net.pf2
        if 0<=i<50:
            self.net.PF.set_state(array([PF1_max*0.,PF2_max*0.]))
        elif 50<=i<100:
            self.net.PF.set_state(array([PF1_max*1.,PF2_max*0.]))
        elif 100<=i<150:
            self.net.PF.set_state(array([PF1_max*1.,PF2_max*1.]))
        elif 150<=i<190:
            self.net.PF.set_state(array([PF1_max*0.,PF2_max*1.]))
        elif 190<=i<200:
            self.net.PF.set_state(array([PF1_max*0.,PF2_max*1.]))
            self.net.CF.set_state(array([1.]))
        elif 200<=i<250:
            self.net.PF.set_state(array([PF1_max*0.,PF2_max*0.]))
            self.net.CF.set_state(array([0.]))
            
    def reset_network(self):
        self.net.reset()
        
    def get_monitor_states(self):
        return self.net.get_monitor_states()

class Network(Network):
    '''
    sets up a network to be simulated
    '''
    
    def __init__(self, w1=.1, w2=0., pf1=.1, pf2=0., mli0=.1, w_cf=0., mli_clamp=False):
        
        self.pf1, self.pf2 = pf1, pf2
        self.mli_clamp = mli_clamp
        
        # define neuron groups
        self.PF = NeuronGroup(2)
        self.MLI = ShuntingNeuronGroup(1, resting_state=array([mli0]), A=1.5, C=1., E=0., F=1., tau=.5)
        self.CF = NeuronGroup(1)

        # connect neurons
        self.C_PF_MLI = PF_MLI_Connection(self.PF,self.MLI,W=array([[w1,w2]]), PF_trace=self.PF)
        self.C_CF_MLI = Connection(self.CF,self.MLI,W=array([[w_cf]]))
        self.MLI.connect(self.C_PF_MLI)
        self.MLI.connect(self.C_CF_MLI)

        # monitor neuron & weight states
        self.PF_monitor = StateMonitor(self.PF)
        self.MLI_monitor = StateMonitor(self.MLI)
        self.W_monitor = StateMonitor(self.C_PF_MLI)
        self.CF_monitor = StateMonitor(self.CF)
    
    def update(self):
        '''
        Update the state of the network
        '''
        if self.mli_clamp:
            self.MLI.set_state(array([0.]))
        else:
            self.MLI.update()
        self.C_PF_MLI.update(beta=.005, CF_active=self.cf_active())
        
    def record(self):
        '''
        record the state of the neurons and weights
        '''
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.W_monitor.record()
        self.CF_monitor.record()
    
    def cf_active(self):
        '''
        returns True if the CF is firing
        '''
        CF = self.CF.get_state()
        if CF[0] > 0:
            return True
        return False
            
    def reset(self):
        '''
        reset the state of the neuron groups and clear monitors
        '''
        self.PF.reset_state()
        self.MLI.reset_state()
        self.PF_monitor.clear_recording()
        self.MLI_monitor.clear_recording()
        self.W_monitor.clear_recording()
        
    def get_monitor_states(self):
        '''
        return recordings from monitors
        '''
        return {'PF':self.PF_monitor.get_recording(),'MLI':self.MLI_monitor.get_recording(),
                'W':self.W_monitor.get_recording()}
```

<h3>Define plotting functions</h3>

```python
def set_axes():
    '''
    format the axis of figures
    '''
    xlabel('time (ms)',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)

def plot_mli(MLI_recording, ls='-', label='MLI Firing rate'):
    '''
    plot the MLI recording
    '''
    plot(MLI_recording[0,:], color='r', ls=ls, lw=2., label=label)
    legend(loc='upper right', fontsize=14)
    set_axes()
    title('MLI Firing rate', fontsize=16)

def plot_pf(PF_recording, W_recording, ind, c='b'):
    '''
    plot the parallel fiber 'ind' recording
    '''
    plot(PF_recording[ind,:], color=c, lw=2., label='PF%s firing rate'%(ind+1))
    plot(W_recording[ind,:], color=c, lw=2., ls='--', label='PF%s-MLI weight (Trial 1)'%(ind+1))
    legend(loc='upper center', fontsize=14)
    set_axes()
    ylim([0,2])
    title('PF%s Firing rate'%ind, fontsize=16)
    
def plot_weight_over_trials(PF_weights_by_trial, ind, c='b'):
    '''
    plot the weight values for PF of index 'ind' over trials
    '''
    plot(PF_weights_by_trial[:,ind], color=c, lw=2., label='PF%s-MLI Weight'%(ind+1))
    legend(loc='lower right', fontsize=14)
    xlabel('Trial',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)
    title('PF Weights', fontsize=16)
    
def plot_weights(pf_weights_by_trial):
    '''
    plot both PF weights over the trials
    '''
    plot_weight_over_trials(pf_weights_by_trial, 0, c='b')
    plot_weight_over_trials(pf_weights_by_trial, 1, c='g')
    
def plot_dashboard(recordings):
    '''
    plot a dashboard of the activity of PFs, MLI, and weights
    '''
    fig = figure(figsize(16,10))
    
    subplot(221)
    plot_mli(recordings[0]['MLI'], ls='--', label='Trial 1')
    plot_mli(recordings[-1]['MLI'], label='Trial 100')
    
    subplot(222)
    pf_weights_by_trial = squeeze([recordings[i]['W'][:,-1] for i in range(100)])
    plot_weights(pf_weights_by_trial)

    subplot(223)
    plot_pf(recordings[0]['PF'], recordings[0]['W'],0)

    subplot(224)
    plot_pf(recordings[0]['PF'], recordings[0]['W'],1,'g')
```

<h2>Experiment #1: PF & MLI Baseline Activity over 100 Trials</h2>

This simulates the changes in synaptic weights when the PF and MLI are firing at baseline.

<h3>Methods</h3>

The network is simulated over 100 consecutive trials. Each trial lasts 250 ms.

In this experiment, $PF_{1}$ has a firing rate of .03 ("3 Hz") between 50 and 150 ms (bottom left panel), and 0.0 otherwise. $MLI_{rest}=.1$. $PF_{2}$ has a firing rate of zero throughout the trial. $w_{1}=.105$ and $w_{2}=0.0$ initially. 

<h3>Results</h3>
We see on trial 1 that the activation of $PF_{1}$ increases the baseline MLI firing rate slightly (top left panel). After 100 trials, the MLI firing rate is the same. We see that the no changes in $w_{1}$ or $w_{2}$ have occurred (top right panel). This is due to $w_{1}$ being in its equilibrium state. For $w_{2}$, there are two reasons it doesn't change; first $PF_{2}$ is zero throughout the trial, and second because $w_{2}$ is zero which means it can only be "activated" by concomitant PF, CF firing.

```python
net = Network(w1=.105, w2=0., pf1=.03, pf2=0., mli0=.1)
sim = Simulation(net, dt=.25)
recordings = []
for trial in range(100):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
plot_dashboard(recordings)
```

<h2>Experiment #2: PF 100 Hz stimulation over 100 Trials</h2>

This simulates the changes in synaptic weights when the PF is stimulated to fire at 100 Hz.

<h3>Methods</h3>

The network is simulated over 100 consecutive trials. Each trial lasts 250 ms.

In this experiment, $PF_{1}$ has a firing rate of 1.0 ("100 Hz") between 50 and 150 ms (bottom left panel), and 0.0 otherwise. $MLI_{rest}=.1$. $PF_{2}$ has a firing rate of zero throughout the trial. $w_{1}=.105$ and $w_{2}=0.0$ initially. 

<h3>Results</h3>
We see on trial 1 that the activation of $PF_{1}$ increases the baseline MLI firing rate significantly (top left panel). After 100 trials, the MLI firing rate is significantly higher. We see $w_{1}$ has increased over the trials but $w_{2}$ has not (top right panel). Since $PF_{1}$ firing is higher, it causes MLI to increase its firing. This increase MLI activity is higher than the equilibrium weight value for this synapse, and thus we see LTP. This is similar to the experiments performed by Smith and Otis (2005) where they showed PF stimulation alone leads to LTP *in vitro*. $w_{2}$ is unchanged for the same reasons as before.

```python
net = Network(w1=.105, w2=0., pf1=1., pf2=0., mli0=.1)
sim = Simulation(net, dt=.25)
recordings = []
for trial in range(100):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
plot_dashboard(recordings)
```

<h2>Experiment #3: Model Activity with MLI clamp over 100 Trials</h2>

This simulates the changes in synaptic weights when the PF is firing at 100 Hz and the MLI is clamped to 0 Hz.

<h3>Methods</h3>

The network is simulated over 100 consecutive trials. Each trial lasts 250 ms.

In this experiment, $PF_{1}$ has a firing rate of 1.0 ("100 Hz") between 50 and 150 ms (bottom left panel), and 0.0 otherwise. $MLI_{rest}=.1$. $PF_{2}$ has a firing rate of zero throughout the trial. $w_{1}=.105$ and $w_{2}=0.0$ initially. 

<h3>Results</h3>
We see that on both the first and last trials the MLI firing rate is zero even in the presence of excitatory input due to the simulated voltage clamp. We see that $w_{1}$ decays quickly toward zero. This can be seen from the learning rule since $MLI < w$, thus LTD. This is similar to the *in vitro* experimental results from Liu and Cull-Candy (2000) and Rancillac and Crepel (2004)

```python
net = Network(w1=.105, w2=0., pf1=1., pf2=0., mli0=.0, mli_clamp=True)
sim = Simulation(net, dt=.25)
recordings = []
for trial in range(100):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
plot_dashboard(recordings)
```

We see that $w_{1}$ decays slowly toward zero for the same reasons as before. It decays slower due to the learning rate being partially determined by $PF_{1}$ activity.

<h2>Experiment #5: Activation of weights due to CF</h2>

This experiment simulates the effect of concomitant PF-CF firing on activating "inactive" synapses.

<h3>Methods</h3>

The network is simulated over 100 consecutive trials. Each trial lasts 250 ms.

In this experiment $w_{1}=w_{2}=0$ initially. $PF_{1}$ has a firing rate of 1.0 ("100 Hz") between 50 and 150 ms (bottom left panel), and 0.0 otherwise. $PF_{2}$ has a firing rate of 1.0 from 100 ms to 200 ms and 0 otherwise (bottom right panel). A CF impinging onto the MLI fires from 190 ms to 200 ms (concomitant with $PF_{2}$).

<h3>Results</h3>
We see that on the first trial the MLI increases its firing rate significantly due to th effect of the CF input and $w_{2}$ increases from 0 to .1 due to the concomitant firing of $PF_{2}$ and the CF. On subsequent trials, $w_{2}$ increases similarly to what we've seen previously.

```python
net = Network(w1=.0, w2=0., pf1=1., pf2=1., mli0=.1, w_cf=1.)
sim = Simulation(net, dt=.25)
recordings = []
for trial in range(100):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
plot_dashboard(recordings)
```
