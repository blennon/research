<h2>Purpose</h2>

The purpose of this experiment is to show the changes in MLI activity and PF-MLI weights over many trials for a simple network composed of three PFs and one MLI. No CF is present.  The learning rule is set to be GSD with PF+CF synaptic activation.

<h2>Results</h2>

* Over 50 trials, both active PF-MLI weights increase and converge to approximately the same quantity.
* The MLI response increases
* The inactive PF-MLI weight stays zero since there is not concomitant PF-CF activation.

```python
from rate_model import *
```

```python
class Simulation(object):
    
    def __init__(self, net, dt=.1):
        self.dt = dt
        self.net = net
    
    def run(self):
        for t in range(250):
            self.set_input_states(t)
            # update state every dt
            for i in range(int(1./self.dt)):
                self.net.update()
            # update monitors every ms
            self.net.record()
    
    def set_input_states(self, i):
        alpha = .5
        if 0<=i<50:
            self.net.PF.set_state(array([0.,0.,0.]))
        elif 50<=i<100:
            self.net.PF.set_state(alpha*array([1.,0.,1.]))
        elif 100<=i<150:
            self.net.PF.set_state(alpha*array([1.,1.,1.]))
        elif 150<=i<190:
            self.net.PF.set_state(alpha*array([0.,1.,0.]))
        elif 190<=i<200:
            self.net.PF.set_state(alpha*array([0.,1.,0.]))
            self.net.CF.set_state(array([0.]))
        elif 200<=i<250:
            self.net.PF.set_state(array([0.,0.,0.]))
            self.net.CF.set_state(array([0.]))
            
    def reset_network(self):
        self.net.reset()
        
    def get_monitor_states(self):
        return self.net.get_monitor_states()
```

```python
w0 = .01
dt = .1
net = PF_MLI_Network(3,1,1,array([0.]),array([[w0,w0,0.]]),array([[1.]]),dt=dt,tau_trace=0.)
sim = Simulation(net, dt)
recordings = []
N_trials = 50
for trial in range(N_trials):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
```

```python
figure(figsize=(8,4))
plot_mli(recordings[0]['MLI'], ls='-.', label='Trial 1')
plot_mli(recordings[-1]['MLI'], label='Trial %s'%N_trials)

figure(figsize=(8,4))
plot(recordings[0]['CF'][0,:], lw=3., color='#6f00ff')
title('CF Firing rate', fontsize=16)
ylim([0.,1.1])
set_axes()

figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],0, '#3366CC')
figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],1,'#339900')
figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],2,'#D65210')
ylim([-.05,2.])

def plot_weights(pf_weights_by_trial):
    plot_weight_over_trials(pf_weights_by_trial, 0, c='#3366CC', ls='-')
    plot_weight_over_trials(pf_weights_by_trial, 1, c='#339900')
    plot_weight_over_trials(pf_weights_by_trial, 2, c='#D65210')

pf_weights_by_trial = squeeze([array([w0,w0,0.])]+[recordings[i]['W'][:,-1] for i in range(N_trials)])
figure(figsize=(8,4))
plot_weights(pf_weights_by_trial)
ylim([-.2,.35])
```
