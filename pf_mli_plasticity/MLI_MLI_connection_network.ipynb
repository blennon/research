```python
def build_pf_mli_connection_matrix(N_PF, N_MLI, spread=1):
    '''
    build the PF-MLI connection matrix.

    N_PF: number of parallel fibers
    N_MLI: number of MLIs
    spread: the number of neighboring neurons that receive the PF input as well

    returns a binary array of shape (N_MLI, N_PF) where a nonzero value
    indicates a connection
    '''
    conv = N_PF/N_MLI
    if float(N_PF)/N_MLI != conv:
        raise Exception('Number of PFs must be an integer multiple of N_MLI')
    W = zeros((N_MLI, N_PF))
    for i in range(N_MLI):
        for j in range(conv):
            begin = max(0, i-spread)
            end = min(N_MLI,i+spread+1)
            W[begin:end,i*conv + j] = 1.
    return W

def build_MLI_MLI_weight_matrix(N_MLI,wmin=.8,wmax=1.,p=.5):
    W = zeros((N_MLI,N_MLI))
    for i in range(N_MLI):
        for j in range(N_MLI):
            if i == j: (wmax-wmin)*rand()+wmin # no autapses
            if rand() < p:
                W[i,j] = (wmax-wmin)*rand()+wmin
    return W
```

```python
class Simulation(object):
    '''
    Simulates the activity of a Network object
    '''
    
    def __init__(self, net, dt=.1):
        '''
        net: Network object
        dt: simulation time step
        '''
        self.dt = dt
        self.net = net
        self.pf_activity = self.create_pf_activity(150)
        
    def create_pf_activity(self, T):
        states = zeros(self.net.PF.get_num_neurons())
        recordings = [states.copy()]
        for i in range(T):
            recordings.append(self.switch_states(states,prob_on=.06).copy())
        return recordings
    
    def run(self):
        for t in range(400):
            # update state every dt
            for i in range(int(1./self.dt)):
                self.set_input_states(t)
                self.net.update()
            # update monitors every ms
            self.net.record()

    def switch_states(self, states, prob_on=.045, prob_off=.035):
        '''
        states: a binary array describing the state of a neuron (firing or not)
        prob_on: probability of switching the state from off to on
        prob_off: probability of switching the state from on to off
        '''
        r = rand(states.shape[0])
        states[(states==0) & (r<prob_on)] = 1 # switch off to on
        states[(states==1) & (r<prob_off)] = 0 # switch on to off
        return states
        
    def set_input_states(self, i):
        if i<250:
            #pf_states = zeros(self.net.PF.get_num_neurons())
            pf_states = self.switch_states(self.net.PF.get_state().copy(),prob_on=.036)
        else:
            #pf_states = self.switch_states(self.net.PF.get_state().copy())
            pf_states = self.pf_activity[i-250]
        self.net.PF.set_state(pf_states)
        #if i < 200 or i > 210:
        #    self.net.CF.set_state(zeros(self.net.CF.get_num_neurons()))
        #else:
        #    self.net.CF.set_state(ones(self.net.CF.get_num_neurons()))
            
    def reset_network(self):
        self.net.reset()
        
    def get_monitor_states(self):
        return self.net.get_monitor_states()

class Network(Network):
    '''
    sets up a network to be simulated
    '''
    
    def __init__(self, w1=.1, w2=0., pf1=.1, pf2=0., mli0=.1, w_cf=2., N_PF=50, N_MLI=10, mli_clamp=False):
        
        self.pf1, self.pf2 = pf1, pf2
        self.mli_clamp = mli_clamp
        
        # define neuron groups
        self.PF = NeuronGroup(N_PF)
        self.MLI = ShuntingNeuronGroup(N_MLI, resting_state=array(mli0*ones(N_MLI)), 
                                       A=1.5, C=1., E=0., F=5., tau=1.)
        self.CF = NeuronGroup(1)

        # connect neurons
        W_PF_MLI = .05*build_pf_mli_connection_matrix(N_PF, N_MLI)
        self.C_PF_MLI = PF_MLI_Connection(self.PF,self.MLI,W=W_PF_MLI, PF_trace=self.PF)
        W_CF_MLI = w_cf*zeros((N_MLI,1))
        self.C_CF_MLI = Connection(self.CF,self.MLI,W=W_CF_MLI)
        W_MLI_MLI = build_MLI_MLI_weight_matrix(N_MLI,wmin=.5,wmax=1.)
        self.C_MLI_MLI = Connection(self.MLI,self.MLI, W_MLI_MLI, type='inhibitory')
        self.MLI.connect(self.C_MLI_MLI)
        self.MLI.connect(self.C_PF_MLI)
        self.MLI.connect(self.C_CF_MLI)

        # monitor neuron & weight states
        self.PF_monitor = StateMonitor(self.PF)
        self.MLI_monitor = StateMonitor(self.MLI)
        self.W_monitor = StateMonitor(self.C_PF_MLI)
        self.CF_monitor = StateMonitor(self.CF)
        
    def update(self):
        '''
        Update the state of the network
        '''
        if self.mli_clamp:
            self.MLI.set_state(array([0.]))
        else:
            self.MLI.update()
        self.C_PF_MLI.update(beta=.005, CF_active=self.cf_active())
        
    def record(self):
        '''
        record the state of the neurons and weights
        '''
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.W_monitor.record()
        self.CF_monitor.record()
    
    def cf_active(self):
        '''
        returns True if the CF is firing
        '''
        CF = self.CF.get_state()
        if CF[0] > 0:
            return True
        return False
            
    def reset(self):
        '''
        reset the state of the neuron groups and clear monitors
        '''
        self.PF.reset_state()
        self.MLI.reset_state()
        self.PF_monitor.clear_recording()
        self.MLI_monitor.clear_recording()
        self.W_monitor.clear_recording()
        
    def get_monitor_states(self):
        '''
        return recordings from monitors
        '''
        return {'PF':self.PF_monitor.get_recording(),'MLI':self.MLI_monitor.get_recording(),
                'W':self.W_monitor.get_recording()}
```

```python
net = Network(w1=.1, w2=0., pf1=1., pf2=0., mli0=.1)
sim = Simulation(net, dt=.25)
recordings = []
for trial in range(100):
    sim.run()
    recordings.append(sim.get_monitor_states())
    sim.reset_network()
```
