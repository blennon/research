```python
class PFMLINetwork(object):
    '''
    PF-MLI network consists of:
        - 3 PFs
        - 1 CF
        - 1 MLI
    '''
    
    def __init__(self):
        '''
        initialize the network
        '''
        # define constants
        self.MLI0 = .25
        self.alpha = 1./(4+self.MLI0)
        self.delta = self.alpha*self.MLI0
        self.beta = .01
        
        # create neuron groups
        self.neuron_groups = {'MLI':NeuronGroup(1, array([self.MLI0])),
                              'PF':NeuronGroup(3),
                              'CF':NeuronGroup(1)}

        # define connections
        self.C_PF_MLI = PF_MLI_Connection(self.neuron_groups['PF'],
                                          self.neuron_groups['MLI'],
                                          W=array([[self.delta, self.delta, 0]]))
        self.C_CF_MLI = Connection(self.neuron_groups['CF'],
                                   self.neuron_groups['MLI'],
                                   array([[1]]))
        self.neuron_groups['MLI'].connect(self.C_PF_MLI)
        self.neuron_groups['MLI'].connect(self.C_CF_MLI)                            
        
        # create monitors
        self.monitors = {'MLI':StateMonitor(self.neuron_groups['MLI']),
                         'PF':StateMonitor(self.neuron_groups['PF']),
                         'CF':StateMonitor(self.neuron_groups['CF']),
                         'C_PF_MLI':StateMonitor(self.C_PF_MLI)}
    
    def reset_network_state(self):
        '''
        reset the state of all neuron groups in the network
        '''
        for ng in self.neuron_groups.values():
            ng.reset_state()
        
    def reset_monitors(self):
        '''
        clear the recordings from all of the monitors
        '''
        for mon in self.monitors.values():
            mon.clear_recording()
            
    def record_monitors(self):
        '''
        records the current state for all the monitors
        '''
        for mon in self.monitors.values():
            mon.record()
        
    def reset_network(self):
        self.reset_network_state()
        self.reset_monitors()
    
    def run_net(self, learn=True):
        '''
        run the network
        '''
        for i in range(250):
            if 50<=i<100:
                self.neuron_groups['PF'].set_state(array([0,1,1]))
            elif 100<=i<150:
                self.neuron_groups['PF'].set_state(array([1,1,1]))
                #self.neuron_groups['PF'].set_state(array([0,1,1]))
            elif 150<=i<190:
                self.neuron_groups['PF'].set_state(array([1,0,0]))
                #self.neuron_groups['PF'].set_state(array([0,0,0]))
            elif 190<=i<200:
                self.neuron_groups['CF'].set_state(array([1]))
            elif i>=200:
                self.neuron_groups['CF'].set_state(array([0]))
                self.neuron_groups['PF'].set_state(array([0,0,0]))
            self.neuron_groups['MLI'].update()
            if learn:
                self.C_PF_MLI.update()
            self.record_monitors()
            
    def get_recordings(self):
        '''
        return the recordings for all monitors in a dictionary
        '''
        return {name:mon.get_recording() for name, mon in self.monitors.iteritems()}
```
