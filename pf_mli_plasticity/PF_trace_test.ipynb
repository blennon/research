```python
from rate_model import *
```

```python
PF = NeuronGroup(2)
PF_trace = PFTrace(2, tau=50.)

PF_mon = StateMonitor(PF)
PF_trace_mon = StateMonitor(PF_trace)
```

```python
for t in range(200):
    if 0<=t<50:
        PF.set_state(array([0.,0.]))
    elif 50<=t<100:
        PF.set_state(array([1.,0.]))
    elif 100<=t<150:
        PF.set_state(array([0.,0.]))
    for dt in range(10):
        PF_trace.update(PF.get_state())
    PF_mon.record()
    PF_trace_mon.record()
```
