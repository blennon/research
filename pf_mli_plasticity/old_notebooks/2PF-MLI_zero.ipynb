<h2>Goals</h2>

The goal of this notebook is to simulate a simple network composed of two PFs and one MLI with the GSD learning rule at the synapses.

The code here is outdated but the graphs are still illustrative.

<h2>Description</h2>

The learning rates are set high so that we can see changes in weights during a single trial. We have the PFs fire in a partially overlapping sequence. We see that the MLI rate increases and decreases as the PF-MLI synaptic weights increase/decrease.

```python
from rate_model import *
```

```python
class Network(Network):
    # define neuron groups
    PF = NeuronGroup(2)
    MLI = ShuntingNeuronGroup(1, resting_state=array([0.0]), A=1., C=1., E=.1, F=1., tau=1.)

    # connect neurons
    C_PF_MLI = PF_MLI_Connection(PF,MLI,W=array([[.1,.1]]),PF_trace=PF)
    MLI.connect(C_PF_MLI)

    # monitor neuron & weight states
    PF_monitor = StateMonitor(PF)
    MLI_monitor = StateMonitor(MLI)
    W_monitor = StateMonitor(C_PF_MLI)
    
    def update(self):
        '''
        Update the state of the network
        '''
        self.MLI.update()
        self.C_PF_MLI.update(beta=.1)
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.W_monitor.record()
        
    def run(self):
        for i in range(350):
            if 0<=i<50:
                self.PF.set_state(array([0.,0.]))
            elif 50<=i<100:
                self.PF.set_state(array([1.,0.]))
            elif 100<=i<150:
                self.PF.set_state(array([1.,1.]))
            elif 150<=i<200:
                self.PF.set_state(array([0.,1.]))
            elif 200<=i<250:
                self.PF.set_state(array([0.,0.]))
            self.update()
```

```python
net = Network()
net.run()
```

```python
def set_axes():
    xlabel('time (ms)',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)
    
figure(figsize=(8,4))
net.MLI_monitor.plot(color='#CC3333', lw=3., label='MLI Firing rate')
legend(loc='upper right', fontsize=14)
set_axes()
title('MLI Firing rate', fontsize=16)

def plot_pf(ind, c='#3366CC'): 
    net.PF_monitor.plot(ind, color=c, lw=3., label='PF%s firing rate'%ind)
    net.W_monitor.plot(ind, color=c, lw=3., ls='--', label='PF%s-MLI weight'%ind)
    legend(loc='upper center', fontsize=14)
    set_axes()
    ylim([0,2])
    title('PF%s Firing rate'%ind, fontsize=16)
    
figure(figsize=(8,4))
plot_pf(0)
figure(figsize=(8,4))
plot_pf(1,'#339900')
```
