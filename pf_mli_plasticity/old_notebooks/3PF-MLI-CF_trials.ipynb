```python
class Network(Network):
    dt = .1 # ms
    
    # define neuron groups
    PF = NeuronGroup(3)
    MLI = ShuntingNeuronGroup(1, A=1.5, resting_state=array([0.1]),tau=1.,dt=dt)
    CF = NeuronGroup(1)

    # connect neurons
    C_PF_MLI = PF_MLI_Connection(PF,MLI,W=array([[.1,.1,0.]]))
    C_CF_MLI = Connection(CF,MLI,W=array([[1.]]))
    MLI.connect(C_PF_MLI)
    MLI.connect(C_CF_MLI)

    # monitor neuron & weight states
    PF_monitor = StateMonitor(PF)
    MLI_monitor = StateMonitor(MLI)
    CF_monitor = StateMonitor(CF)
    W_monitor = StateMonitor(C_PF_MLI)
    
    def cf_active(self):
        '''
        returns True if the CF is firing
        '''
        CF = self.CF.get_state()
        if CF[0] > 0:
            return True
        return False
    
    def update(self):
        '''
        Update the state of the network
        '''
        self.MLI.update()
        self.C_PF_MLI.update(beta=.0005, CF_active=self.cf_active())
    
    def record(self):
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.CF_monitor.record()
        self.W_monitor.record()
    
    def set_input_states(self, i):
        if 0<=i<50:
            self.PF.set_state(array([0.,0.,0.]))
        elif 50<=i<100:
            self.PF.set_state(array([1.,0.,1.]))
        elif 100<=i<150:
            self.PF.set_state(array([1.,1.,1.]))
        elif 150<=i<190:
            self.PF.set_state(array([0.,1.,0.]))
        elif 190<=i<200:
            self.PF.set_state(array([0.,1.,0.]))
            self.CF.set_state(array([1.]))
        elif 200<=i<250:
            self.PF.set_state(array([0.,0.,0.]))
            self.CF.set_state(array([0.]))
            
    def run(self):
        for t in range(250):
            self.set_input_states(t)
            # update state every dt
            for i in range(int(1./self.dt)):
                self.update()
            # update monitors every ms
            self.record()
            
    def reset(self):
        '''
        reset the state of the neuron groups and clear monitors
        '''
        self.PF.reset_state()
        self.MLI.reset_state()
        self.PF_monitor.clear_recording()
        self.MLI_monitor.clear_recording()
        self.W_monitor.clear_recording()
        self.CF_monitor.clear_recording()
        
    def get_monitor_states(self):
        '''
        return recordings from monitors
        '''
        return {'PF':self.PF_monitor.get_recording(),'MLI':self.MLI_monitor.get_recording(),
                'W':self.W_monitor.get_recording(), 'CF':self.CF_monitor.get_recording()}
```

```python
net = Network()
recordings = []
for trial in range(100):
    net.run()
    recordings.append(net.get_monitor_states())
    net.reset()
```

```python
def set_axes():
    xlabel('time (ms)',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)

def plot_mli(MLI_recording, ls='-', label='MLI Firing rate'):
    plot(MLI_recording[0,:], color='r', ls=ls, lw=2., label=label)
    legend(loc='upper center', fontsize=14)
    set_axes()
    title('MLI Firing rate', fontsize=16)

def plot_pf(PF_recording, W_recording, ind, c='b'): 
    plot(PF_recording[ind,:], color=c, lw=2., label='PF%s firing rate'%ind)
    plot(W_recording[ind,:], color=c, lw=2., ls='--', label='PF%s-MLI weight'%ind)
    legend(loc='upper center', fontsize=14)
    set_axes()
    ylim([0,2])
    title('PF%s Firing rate'%ind, fontsize=16)
    
def plot_figures(MLI_recording,PF_recording, W_recording):
    figure(figsize=(8,4))
    plot_mli(MLI_recording)
    figure(figsize=(8,4))
    plot_pf(PF_recording, W_recording,0)
    figure(figsize=(8,4))
    plot_pf(PF_recording, W_recording,1,'g')
    
def plot_figures(MLI_recording,PF_recording, W_recording):
    figure(figsize=(8,4))
    plot_mli(MLI_recording)
    figure(figsize=(8,4))
    plot_pf(PF_recording, W_recording,0)
    figure(figsize=(8,4))
    plot_pf(PF_recording, W_recording,1,'g')
    
def plot_weight_over_trials(PF_weights_by_trial, ind, c='b'): 
    plot(PF_weights_by_trial[:,ind], color=c, lw=2., label='PF%s-MLI Weight'%ind)
    legend(loc='lower right', fontsize=14)
    xlabel('Trial',fontsize=14)
    xticks(fontsize=14)
    yticks(fontsize=14)
    title('PF Weights', fontsize=16)
    
def plot_weights(pf_weights_by_trial):
    plot_weight_over_trials(pf_weights_by_trial, 0, c='b')
    plot_weight_over_trials(pf_weights_by_trial, 1, c='g')
    plot_weight_over_trials(pf_weights_by_trial, 2, c='k')
```

```python
figure(figsize=(8,4))
plot_mli(recordings[0]['MLI'], ls='--', label='Trial 1')
plot_mli(recordings[-1]['MLI'], label='Trial 100')

figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],0)
figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],1,'g')
figure(figsize=(8,4))
plot_pf(recordings[0]['PF'], recordings[0]['W'],2,'k')
```
