<h2>Purpose</h2>
The purpose of this experiment is to test the effect of the PF trace on learning when a CF arrives a few ms after the end of the PF burst.

<h2>Results</h2>

* The trace enables the weight to predict rewards (CF signal) that arrive after the end of the PF stimulus
* The longer the trace, the lower the weight value, since the PF trace gates learning and there is now more learning during times with no MLI activity

```python
from rate_model import *
```

```python
class Simulation(object):
    
    def __init__(self, net, dt=.1):
        self.dt = dt
        self.net = net
    
    def run(self):
        for t in range(250):
            self.set_input_states(t)
            # update state every dt
            for i in range(int(1./self.dt)):
                self.net.update()
            # update monitors every ms
            self.net.record()
    
    def set_input_states(self, i):
        if 0<=i<50:
            self.net.PF.set_state(array([0.,0.,0.]))
        elif 50<=i<100:
            self.net.PF.set_state(array([1.,0.,0.]))
        elif 100<=i<150:
            self.net.PF.set_state(array([1.,0.,0.]))
        elif 150<=i<160:
            self.net.PF.set_state(array([0.,0.,0.]))
        elif 160<=i<170:
            self.net.CF.set_state(array([1.]))
        elif 170<=i<200:
            self.net.CF.set_state(array([0.]))
        elif 200<=i<250:
            self.net.PF.set_state(array([0.,0.,0.]))
            self.net.CF.set_state(array([0.]))
            
    def reset_network(self):
        self.net.reset()
        
    def get_monitor_states(self):
        return self.net.get_monitor_states()
```

```python
results = {}
for w in [0.,1.]:
    for tau in [0., 10.]:      
        dt = .1
        net = PF_MLI_Network(3,1,1,array([.1]),array([[.1,.1,0.]]),array([[w]]),dt=dt,tau_trace=tau)
        sim = Simulation(net, dt)
        recordings = []
        N_trials = 50
        for trial in range(N_trials):
            sim.run()
            recordings.append(sim.get_monitor_states())
            sim.reset_network()
        results[(w,tau)] = recordings
```

<h2>Comparison of weight values</h2>

```python
for param, recordings in results.items():
    w,tau = param
    print 'CF weight = %s, tau = %s, PF final weight = %s' % (w,tau,recordings[-1]['W'][0,-1])
```

We can see, in the case where there is a trace, i.e. tau > 0, the final PF-MLI weight is higher when CF is present (CF>0). In the case where there is no trace, the final PF weight is the same, regardless of CF or not
