```python
from rate_model import *
```

```python
class Network(object):
    # define neuron groups
    PF = NeuronGroup(1)
    MLI = ShuntingNeuronGroup(1, A=1.5, resting_state=array([0.1]))

    # connect neurons
    C_PF_MLI = PF_MLI_Connection(PF,MLI,W=array([[.10]]))
    MLI.connect(C_PF_MLI)

    # monitor neuron & weight states
    PF_monitor = StateMonitor(PF)
    MLI_monitor = StateMonitor(MLI)
    W_monitor = StateMonitor(C_PF_MLI)
    
    def update(self):
        '''
        Update the state of the network
        '''
        self.MLI.update()
        self.C_PF_MLI.update(beta=.1)
        self.PF_monitor.record()
        self.MLI_monitor.record()
        self.W_monitor.record()
```

```python
def run(T):
    for i in range(T):
        net.PF.set_state(array([1.]))
        net.update()
```

```python
net = Network()
run(100)
```

```python
figure(figsize=(8,4))
net.MLI_monitor.plot(label='MLI Firing rate')
net.W_monitor.plot(label='PF-MLI weight')
legend(loc='lower right', fontsize=14)
xlabel('time (ms)',fontsize=14)
xticks(fontsize=14)
yticks(fontsize=14)
title('MLI Firing rate and weight', fontsize=14)
```
