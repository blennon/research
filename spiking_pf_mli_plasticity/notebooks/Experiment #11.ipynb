<h1>Experiment #11</h1>
This experiment simulates a single MLI isolated from inhibitory synaptic current that fires spontaneously. The PF-MLI weight begins near its equilibrium value. The MLI fires spontaneously throughout the entire experiment. During the first 5 seconds, the PF fires at its baseline rate of .33 Hz. A single parallel fiber is then stimulated to fire at 10 Hz (according to Poisson statistics) 

```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../../")
from brian import *

from neuron_models import *

from spiking_pf_mli_plasticity import *
import seaborn as sns
```

```python
save_results = True
exper_dir = '/home/bill/research/data/pf_mli_learning/experiment_11/'

N_GR = 10
eq_time = 5*second
trial_time = 1*second
stim_len = 1*second
base_rate = .33*Hz
stim_rate = 2*Hz
wmin = 0.2
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    
    # set up the network
    pf_rates = build_pf_rates_func(eq_time, trial_time, N_GR, base_rate, stim_rate, stim_len)
    GR, MLI, S_GR_MLI = setup_independent_isolated_mli_nets(N_GR, initial_weights, pf_rates, wmin)

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        
    # monitors
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W = setup_monitors(GR, MLI, S_GR_MLI)
    
    # plasticity
    @network_operation(defaultclock)
    def learn():
        update_weights(S_GR_MLI, GR_R, MLI_R, wmin)
        
    run(1*65*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(.22*ones(N_GR))

if save_results:
    objects_dict = {'GR_S':GR_S, 'GR_R':GR_R, 'MLI_V':MLI_V, 
                'MLI_S':MLI_S,'MLI_R':MLI_R, 'W':W}
    pickle_results(objects_dict, exper_dir)
```
