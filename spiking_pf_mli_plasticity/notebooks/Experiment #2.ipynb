<h1>Experiment #2</h1>
This experiment simulates a single MLI isolated from inhibitory synaptic current that fires spontaneously. The PF-MLI weights begin near their equilibrium values. The experiment begins with a five second baseline period where PFs and MLIs fire near their baseline values. Baseline GR firing rate is .33 Hz. A bundle of parallel fibers is then stimulated to produce 100 Hz bursts (according to Poisson statistics) in PFs for 100 ms every second. 

```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
save_results = True
exper_dir = '/home/bill/research/data/pf_mli_learning/experiment_2/'

N_GR = 8
eq_time = 5*second
trial_time = 1*second
stim_len = 100*ms
base_rate = .33*Hz
stim_rate = 100*Hz
wmin = 0.2
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    
    # set up the network
    pf_rates = build_pf_rates_func(eq_time, trial_time, N_GR, base_rate, stim_rate, stim_len)
    GR, MLI, S_GR_MLI = setup_isolated_mli_net(N_GR, initial_weights, pf_rates, wmin)

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        
    # monitors
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W = setup_monitors(GR, MLI, S_GR_MLI)
    
    # plasticity
    @network_operation(defaultclock)
    def learn():
        update_weights(S_GR_MLI, GR_R, MLI_R)
        S_GR_MLI.v[:,:] = wmin + (1-wmin)*S_GR_MLI.w[:,:]
        
    run(1*30*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(.22*ones(N_GR))

if save_results:
    objects_dict = {'GR_S':GR_S, 'GR_R':GR_R, 'MLI_V':MLI_V, 
                'MLI_S':MLI_S,'MLI_R':MLI_R, 'W':W}
    pickle_results(objects_dict, exper_dir)
```
