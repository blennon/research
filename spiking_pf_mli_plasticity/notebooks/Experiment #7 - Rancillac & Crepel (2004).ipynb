<h1>Experiment #7 - Rancillac & Crepel (2004)</h1>
This experiment attempts to reproduce the results of R&C (2004) where a depolarization in MLI results in LTP.

This experiment simulates a single MLI isolated from inhibitory synaptic current that begins in voltage clamp @ -60 mV. The PF-MLI weights begin at their minimum values.  It is assumed that during the induction period of the R&C experiment, the synaptic efficacy decreases. The PF is stimulated at 2 Hz while the MLI is depolarized to 0 mV. We simulate this 0 mV depolarization as a fixed firing rate of 50 Hz.

```python
%load_ext autoreload
%autoreload 2
import sys

sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
import seaborn as sns
```

```python
class MLI_R_constant(object):
    @check_units(rate=hertz)
    def __init__(self, MLI, rate):
        self.rate = rate
        self.MLI = MLI

    @check_units(mli_max=hertz)
    def get_normalized_firing_rates(self,mli_max):
        return float(self.rate/mli_max)*ones(len(self.MLI))
    
    def __len__(self):
        return len(self.MLI)
```

```python
save_results = True
exper_dir = '/home/bill/research/data/pf_mli_learning/experiment_7/'

N_GR = 8
eq_time = 5*second
trial_time = 1*second
stim_len = 1*second
base_rate = .33*Hz
stim_rate = 2*Hz
wmin = 0.2
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    
    # set up the network
    pf_rates = build_pf_rates_func(eq_time, trial_time, N_GR, base_rate, stim_rate, stim_len)
    GR, MLI, S_GR_MLI = setup_isolated_mli_net(N_GR, initial_weights, pf_rates, wmin)
    
    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        if defaultclock.t >= eq_time:
            MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
            MLI.I += 0.01 * nA
        else:
            MLI.V = -60 * mV
            
    # monitors
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W = setup_monitors(GR, MLI, S_GR_MLI)
    #Constant_MLI_R = MLI_R_constant(MLI, 50*Hz)
    
    # plasticity
    @network_operation(defaultclock)
    def learn():
        update_weights(S_GR_MLI, GR_R, MLI_R, wmin)
        
    run(1*65*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(0.*ones(N_GR))

if save_results:
    objects_dict = {'GR_S':GR_S, 'GR_R':GR_R, 'MLI_V':MLI_V, 
                'MLI_S':MLI_S, 'W':W}
    pickle_results(objects_dict, exper_dir)
```
