The purpose of this notebook is to test the viability of using a low-pass filtered version of the MLI membrane potential in the learning rule.

```python
import sys
sys.path.append("../../")
from brian import *

from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
def run_net():
    
    defaultclock.dt = .25*ms
    MLI = MLIGroup(1)
    
    record_clock = Clock(1*ms)
    MLI_V, tau_MLI_V = [MLI.V.copy()], 50*ms
    MLI_SV = StateMonitor(MLI, 'V', record=True, clock=record_clock)
    MLI_S = SpikeMonitor(MLI)
    MLI_R = RealTimeRateMonitor(MLI, tau_f=60*ms, tau_r=15*ms, record=True, record_clock=record_clock)
    
    
    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        MLI.I += .02 * nA * (defaultclock.t / 5*second)
        if len(MLI.get_spikes()>0):
            MLI_V.append(MLI_V[-1] + defaultclock.dt*(20*mV - MLI_V[-1])/tau_MLI_V)
        else:
            MLI_V.append(MLI_V[-1] + defaultclock.dt*(MLI.V - MLI_V[-1])/tau_MLI_V)
        
    run(5*second)
    
    return MLI_V, MLI_SV, MLI_S, MLI_R
```

```python
MLI_V, MLI_SV, MLI_S, MLI_R = run_net()
```

```python
MLI_SV.insert_spikes(MLI_S)
```
