```python
%load_ext autoreload
%autoreload 2
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
set_global_preferences(useweave=True, usenewpropagate=True, usecodegen=True, usecodegenweave=True)
import seaborn as sns
import datetime
import cPickle
```

```python
N_GR = 5
stim_rate = 100*Hz
wmin = 0.2
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms

    GR = PoissonGroup(N_GR, rates=stim_rate*ones(N_GR))
    MLI = MLIGroup(N_GR)
    S_GR_MLI = Synapses(GR,MLI,model='''w:1
                        v:1''',pre='g_ampa_fast+=MLI.g_ampa_*v; g_ampa_slow+=MLI.g_ampa_*v; n+=1')
    S_GR_MLI[0,0] = 1.
    S_GR_MLI[1,0] = 1.
    S_GR_MLI[2,2] = 1.
    S_GR_MLI[3,3] = 1.
    S_GR_MLI[4,4] = 1.
    S_GR_MLI.w[:,:] = initial_weights
    S_GR_MLI.v[initial_weights>0] = wmin + (1-wmin)*S_GR_MLI.w[initial_weights>0]
    CF = PoissonGroup(1, rates=10*Hz)
    S_CF_MLI = Synapses(CF,MLI,model='''w:1''')
    S_CF_MLI[0,1] = True
    S_CF_MLI[0,2] = True
    S_CF_MLI.w[:,:] = 1.

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        MLI.I[N_GR-1] = -60*mV
        
    # monitors
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W = setup_monitors(GR, MLI, S_GR_MLI)
    CF_R = RealTimeRateMonitor(CF, tau_f=60*ms, tau_r=15*ms, record=True, record_clock=Clock(1*ms))
    
    # plasticity
    @network_operation(defaultclock)
    def learn():
        update_weights_cf(S_GR_MLI, S_CF_MLI, GR_R, MLI_R, CF_R, wmin,beta=.003)
        
    run(1*2*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(array([0.,0.,0.,.22,.22]))
```
