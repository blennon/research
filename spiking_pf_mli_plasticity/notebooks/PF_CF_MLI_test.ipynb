This notebook tests the implementation of PF-MLI learning when CF are present.

```python
%load_ext autoreload
%autoreload 2
import datetime
import os
from brian import *
import sys
sys.path.append('../../')
from neuron_models import *
from spiking_pf_mli_plasticity import *
import cPickle
import time
set_global_preferences(useweave=True, usenewpropagate=True, usecodegenweave=True)
defaultclock.dt = .25*ms
```

```python
inact_to_active = (0,0) # inactive synapse -- will show activation of this synapse
active_to_inactive = (6,1) # active synapse -- will show inactivation of this synapse
active_str = (6,2) # active synapse -- will show strengthing of this synapse
active_weak = (6,3) # active synapse -- will show weakening of this synapse
```

```python
T = 400*msecond
defaultclock.dt = .25*ms
wmin = .2
N_CF = 2

GR, CF, MLI, S_GR_MLI, S_CF_MLI = setup_mli_net(100*Hz,10*Hz,wmin=wmin, N_CF=N_CF,N_GR_clusters=5)
GR_S, GR_R, MLI_V, MLI_S, MLI_R, CF_S, CF_R, W = setup_monitors_cf(GR,MLI,CF,S_GR_MLI)

S_GR_MLI.w[active_to_inactive] = .05*wmin
S_GR_MLI.v[active_to_inactive] = wmin + (1-wmin)*.05*wmin
S_CF_MLI.w[0,1] = 0.
S_CF_MLI.w[1,1] = 0.
S_CF_MLI.w[0,3] = 0.
S_CF_MLI.w[1,3] = 0.

@network_operation(Clock(dt=defaultclock.dt))
def random_current():
    MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
    MLI.V[1] = -70*mV # check this works
    MLI.V[3] = -70*mV # check this works
    
# plasticity
@network_operation(Clock(dt=1*ms))
def learn():
    update_weights_cf(S_GR_MLI, S_CF_MLI, GR_R, MLI_R, CF_R, wmin, beta=.004)

start_vals = [S_GR_MLI.v[inact_to_active], S_GR_MLI.v[active_to_inactive], 
              S_GR_MLI.v[active_str], S_GR_MLI.v[active_weak], S_GR_MLI.v[:,:].shape[0]]

start = time.time()
run(T)
print time.time() - start

end_vals = [S_GR_MLI.v[inact_to_active], S_GR_MLI.v[active_to_inactive], 
            S_GR_MLI.v[active_str], S_GR_MLI.v[active_weak], S_GR_MLI.v[:,:].shape[0]]
```

```python
assert start_vals[0] == 0.
assert end_vals[0] > start_vals[0]

assert start_vals[1] > 0.
assert end_vals[1] == 0.

assert end_vals[2] > start_vals[2]

assert end_vals[3] < start_vals[3]

assert end_vals[4] == start_vals[4]
```
