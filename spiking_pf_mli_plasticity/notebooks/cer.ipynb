```python
T_trial = 1400 * ms
T_CS = 650 * ms
D_CS = 350 * ms
D_US = 100 * ms
N_trials = 10
N_GR = 100
N_MLI = 10

    
# moved to stimulus.py

```

```python
defaultclock.dt = .25*ms

# precompute gr rates
p_up_to_down = switch_prob(100*ms, 1*ms)
p_down_to_up = switch_prob(450*ms, 1*ms)
gr_rates = GR_rates(T_trial,T_CS,D_CS,defaultclock.dt,
                    N_GR,.5*Hz,100*Hz,p_down_to_up,p_up_to_down)

# setup MLIs
MLI = MLIGroup(N_MLI)
S_MLI_MLI = Synapses(MLI,MLI,model='w:1',pre='g_inh+=MLI.g_inh_*w')
S_MLI_MLI.connect_random(sparseness=.4)
S_MLI_MLI.w[:,:] = 1.#'rand()'

# setup GRs
GR = PoissonGroup(N_GR,rates=gr_rates)
S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa_fast+=MLI.g_ampa_*w; g_ampa_slow+=MLI.g_ampa_*w')
S_GR_MLI.connect_random(sparseness=.35)
S_GR_MLI.w[:,:] = '.1 + .2*rand()'

#CF = SpikeGeneratorGroup(1,[(0,(i*ms+j*T_trial)) for j in range(N_trials) for i in range(650,700,10)])
#S_CF_MLI = Synapses(CF,MLI,model='w:1',pre='g_ampa_fast+=MLI.g_ampa_*w; g_ampa_slow+=MLI.g_ampa_*w')
#S_CF_MLI[0,5] = 1.
#S_CF_MLI.w[:,:] = 3 # lookup physio value

@network_operation(Clock(dt=defaultclock.dt))
def random_current():
    '''inject current into MLIs'''
    MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA

# set up state monitors
record_clock = Clock(1*ms)
GR_S = SpikeMonitor(GR)
GR_R = RealTimeRateMonitor(GR, record=True, record_clock=record_clock, tau_f=20*ms)
MLI_V = StateMonitor(MLI, 'V', record=True, clock=record_clock)
MLI_S = SpikeMonitor(MLI)
MLI_R = RealTimeRateMonitor(MLI, tau_f=50*ms, tau_r=10*ms, record=True, record_clock=record_clock)
W = StateMonitor(S_GR_MLI,'w',record=True, clock=record_clock)

# PF-MLI learning
beta = .005
syn_inds, pre_inds = synapse_inds(range(N_GR),S_GR_MLI,'pre')
_, post_inds = synapse_inds(range(N_MLI),S_GR_MLI,'post')
@network_operation(defaultclock)
def update_weights():
    
    raise NotImplementedError("need to call get_normalized_firing_rates() method from RateMonitor")
    #mli_fr = calc_normal_firing_rates(MLI_R, 100)
    #gr_fr = calc_normal_firing_rates(GR_R, 500)    
    S_GR_MLI.w[:,:] += beta*gr_fr[pre_inds]*(mli_fr[post_inds] - 2.*S_GR_MLI.w[:,:])
        
run(N_trials*T_trial)
```
