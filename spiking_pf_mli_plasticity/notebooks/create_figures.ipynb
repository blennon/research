The purpose of this notebook is to plot the data from multiple experiments to create the figures appearing in the manuscript

```python
import cPickle
import seaborn as sns
from pandas import *
```

```python
def plot_weights_by_trial(Ws, trial_len, trial_start_time, dt):
    '''
    Plot the value of the weight at the end of each trial. Also show
    the starting weight value as the 0th trial (i.e. during equilibrium
    period).

    fig: matplotlib figure object
    W: weight monitor object
    trial_len: duration of each trial in seconds
    trial_start_time: time when the first
    '''
    # style
    sns.set_style("whitegrid")
    sns.set_context("poster")

    start_ind = int(trial_start_time/dt)
    # use the first value of each trial as the value at end of previous trial
    Wts = []
    for W in Ws:
        wt = W[:,start_ind:][:,::int(trial_len/dt)]
        wt = hstack((wt,W[:,-1][...,None])) # this adds the last value of the last trial to be the end value of last trial
        w0 = wt[:,0][...,None] # initial values
        deltaw = (wt - w0)/w0
        Wts.append(deltaw)
    trials = range(wt.shape[1])
    sns.tsplot(dstack(Wts), time=trials, ci=100, linewidth=3)
    ylabel(r'$\frac{\Delta w}{w_{0}}$', fontsize=35, rotation='horizontal',labelpad=25)
    xlabel('Trial number')
```

```python
with open('/home/bill/research/data/pf_mli_learning/experiment_1/12-03-2014_16:49/W.pkl') as f:
    exp1_W = cPickle.load(f)
    
with open('/home/bill/research/data/pf_mli_learning/experiment_10/12-03-2014_16:51/W.pkl') as f:
    exp10_W = cPickle.load(f)
    
with open('/home/bill/research/data/pf_mli_learning/experiment_11/12-03-2014_16:51/W.pkl') as f:
    exp11_W = cPickle.load(f)
    
with open('/home/bill/research/data/pf_mli_learning/experiment_3/12-03-2014_16:51/W.pkl') as f:
    exp3_W = cPickle.load(f)
```

```python
fig = figure(figsize=(20,10))
plot_weights_by_trial([exp1_W, exp3_W, exp10_W, exp11_W], 1.,5.,.001)
```
