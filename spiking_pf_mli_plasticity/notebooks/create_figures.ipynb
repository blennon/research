The purpose of this notebook is to plot the data from multiple experiments to create the figures appearing in the manuscript

```python
import cPickle
import seaborn as sns
from pandas import *
```

```python
def plot_weights_by_trial(Ws, trial_len, trial_start_time, dt, ax=None, palette=None):
    '''
    Plot the value of the weight at the end of each trial. Also show
    the starting weight value as the 0th trial (i.e. during equilibrium
    period).

    fig: matplotlib figure object
    W: weight monitor object
    trial_len: duration of each trial in seconds
    trial_start_time: time when the first
    '''
    # style
    sns.set_style("whitegrid")
    sns.set_context("poster")

    if palette is None:
        palette = sns.color_palette()
        
    start_ind = int(trial_start_time/dt)
    # use the first value of each trial as the value at end of previous trial
    Wts = []
    for W in Ws:
        wt = W[:,start_ind:][:,::int(trial_len/dt)]
        wt = hstack((wt,W[:,-1][...,None])) # this adds the last value of the last trial to be the end value of last trial
        #w0 = wt[:,0][...,None] # initial values
        w0 = W[:,0][...,None]
        deltaw = (wt - w0)/w0
        Wts.append(deltaw)
    trials = range(wt.shape[1])
    if ax is not None:
        sns.tsplot(dstack(Wts), time=trials, ci=100, linewidth=3, ax=ax, color=palette,err_kws={"alpha": .35})
    else:
        sns.tsplot(dstack(Wts), time=trials, ci=100, linewidth=3, color=palette,err_kws={"alpha": .35})
    #ylabel(r'$\frac{\Delta w}{w_{0}}$', fontsize=35, rotation='horizontal',labelpad=25)
    ylabel('percent weight change',fontsize=20)
```

```python
def format_barplot_data(experiments):
    exper_numbers = []
    weights = []

    for i,e in enumerate(experiments):
        W = e[1]
        final_norm_weights = W[:,-1]/W[:,0]
        weights.append(final_norm_weights)
        #exper_numbers.append((i+1)*ones_like(final_norm_weights))
        exper_numbers.append([e[0] for _ in range(final_norm_weights.shape[0])])
    return hstack(exper_numbers), hstack(weights)
```
