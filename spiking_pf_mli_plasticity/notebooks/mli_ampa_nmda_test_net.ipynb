```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *



import seaborn as sns
```

```python
N_GR = 20
eq_time = 1*second
trial_time = 1*second
stim_len = 100*ms
base_rate = .33*Hz
stim_rate = 500*Hz
wmin = 0.2
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    
    # set up the network
    pf_rates = build_pf_rates_func(eq_time, trial_time, N_GR, base_rate, stim_rate, stim_len)
    GR, MLI, S_GR_MLI = setup_isolated_mli_net(N_GR, initial_weights, pf_rates,wmin)

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        
    # monitors
    mclock = Clock(dt=1*ms)
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W = setup_monitors(GR, MLI, S_GR_MLI)
    MLI_N = StateMonitor(MLI,'gv_nmda',record=True,clock=mclock)
    MLI_A = StateMonitor(MLI, 'g_ampa', record=True,clock=mclock)
    MLI_n = StateMonitor(MLI, 'n', record=True,clock=mclock)
    
    
    # plasticity
    @network_operation(defaultclock)
    def learn():
        update_weights(S_GR_MLI, GR_R, MLI_R)
        S_GR_MLI.v[:,:] = wmin + (1-wmin)*S_GR_MLI.w[:,:]
        
    run(1*2*second)
    
    return MLI_N, MLI_A, MLI_n, GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
M_GR_gvnmda, M_GR_ampa, MLI_n, GR_S, GR_R, M_GR_V, MLI_S, MLI_R, W, w = run_net(array([.22]))
```
