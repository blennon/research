```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
eq_time = 5*second
trial_time = 1*second
    
def pf_rates(t):
    if t>=eq_time and t%(trial_time)<100*ms:
        return 100*Hz
    return 2*Hz
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    GR = PoissonGroup(1,rates=pf_rates)
    MLI = MLIGroup(1)
    S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa_fast+=MLI.g_ampa_*w; g_ampa_slow+=MLI.g_ampa_*w')
    S_GR_MLI[:,:] = 1
    S_GR_MLI.w[:,:] = initial_weights

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        if defaultclock.t < eq_time:
            MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        else:
            MLI.V = -60*mV
        
    record_clock = Clock(1*ms)
    GR_S = SpikeMonitor(GR)
    GR_R = RealTimeRateMonitor(GR, record=True, record_clock=record_clock)
    MLI_V = StateMonitor(MLI, 'V', record=True, clock=record_clock)
    MLI_S = SpikeMonitor(MLI)
    MLI_R = RealTimeRateMonitor(MLI, tau_f=60*ms, tau_r=15*ms, record=True, record_clock=record_clock)
    W = StateMonitor(S_GR_MLI,'w',record=True, clock=record_clock)
    
    beta = .001
    @network_operation(defaultclock)
    def update_weights():
        mli_fr = min(MLI_R.get_firing_rates(),100.)/100.
        gr_fr = GR_R.get_firing_rates()
        gr_fr[gr_fr>500.] = 500.
        gr_fr /= 500.
        delta_w = beta*gr_fr*(mli_fr - S_GR_MLI.w[:,:])
        S_GR_MLI.w[:,:] += delta_w
        
    run(15*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(.3)
```

```python
# plot MLI membrane potential trace
figure(figsize(20,6))
MLI_V.insert_spikes(MLI_S)
MLI_V.plot()
title('Membrane potential over time', fontsize=18)
xlabel('time (s)', fontsize=16)
ylabel('Membrane Potential (V)', fontsize=16)

# plot MLI firing rate trace
figure(figsize(20,6))
plot(MLI_R.getvalues())
title('MLI Firing Rate', fontsize=18)
xlabel('time (ms)', fontsize=16)
ylabel('Firing Rate (Hz)', fontsize=16)

# plot GR firing rate vtraces from last trial
figure(figsize(20,6))
gr_frs = squeeze(GR_R.getvalues())
plot(gr_frs)
title('GR Firing Rate', fontsize=18)
xlabel('time (ms)', fontsize=16)
ylabel('Firing Rate (Hz)', fontsize=16)

# plot the weight value
figure(figsize(20,6))
plot(squeeze(W.getvalues()))
title('Synaptic Weight', fontsize=18)
xlabel('time (ms)', fontsize=16)
ylabel('Synaptic Weight', fontsize=16)
```
