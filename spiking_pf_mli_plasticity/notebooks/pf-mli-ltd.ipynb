```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
eq_time = 5*second
trial_time = 1*second
    
def pf_rates(t):
    if t>=eq_time and t%(trial_time)<100*ms:
        return 100*Hz
    return 2*Hz
        
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    GR = PoissonGroup(1,rates=pf_rates)
    MLI = MLIGroup(1)
    S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa_fast+=MLI.g_ampa_*w; g_ampa_slow+=MLI.g_ampa_*w')
    S_GR_MLI[:,:] = 1
    S_GR_MLI.w[:,:] = initial_weights

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        if defaultclock.t < eq_time:
            MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        else:
            MLI.V = -60*mV
        
    record_clock = Clock(1*ms)
    GR_S = SpikeMonitor(GR)
    GR_R = RealTimeRateMonitor(GR, record=True, record_clock=record_clock)
    MLI_V = StateMonitor(MLI, 'V', record=True, clock=record_clock)
    MLI_S = SpikeMonitor(MLI)
    MLI_R = RealTimeRateMonitor(MLI, tau_f=60*ms, tau_r=15*ms, record=True, record_clock=record_clock)
    W = StateMonitor(S_GR_MLI,'w',record=True, clock=record_clock)
    
    beta = .0015
    MLI_max, GR_max = 150., 500
    @network_operation(defaultclock)
    def update_weights():
        mli_fr = min(MLI_R.get_firing_rates(),MLI_max)/MLI_max
        gr_fr = GR_R.get_firing_rates()
        gr_fr[gr_fr>GR_max] = GR_max
        gr_fr /= GR_max
        delta_w = beta*gr_fr*(mli_fr - S_GR_MLI.w[:,:])
        S_GR_MLI.w[:,:] += delta_w
        
    run(15*second)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, w = run_net(.3)
```

```python
end_time = float(defaultclock.t)
T = linspace(0,end_time,1000*end_time)

gs = GridSpec(3,1,height_ratios=[2,1,3])

# plot MLI membrane potential trace
fig = figure(figsize(25,14))
ax1 = fig.add_subplot(gs[0])
MLI_V.insert_spikes(MLI_S)
ax1.plot(T,squeeze(MLI_V.getvalues())*1000, color='#8C2318')
ax1.set_xlim(xmax=end_time)
ax1.tick_params(labelsize=16)
ax1.set_ylabel('MLI Membrane Potential (mV)', fontsize=16)
ax1.set_yticks(range(-80,20,20))
ax1.set_xticks(range(0,16,1))
simpleaxis(ax1)

# plot MLI firing rate trace
ax2 = fig.add_subplot(gs[1])
ax2.plot(T,MLI_R.getvalues(), color='#8C2318', linewidth=1.5)
ax2.set_xlim(xmax=end_time)
ax2.tick_params(labelsize=16)
ax2.set_yticks(range(0,50,10))
ax2.set_ylabel('MLI Firing Rate (Hz)', fontsize=16)
ax2.set_xticks(range(0,16,1))
simpleaxis(ax2)

# Overlay plot of GR firing rate and synapse weight
ax1 = fig.add_subplot(gs[2])
gr_frs = squeeze(GR_R.getvalues())
ax1.plot(T,gr_frs, color='#0B486B')
ax1.set_ylabel('GR Firing Rate (Hz)', color='#0B486B', fontsize=20)
ax1.tick_params(labelsize=16)
ax1.spines['top'].set_visible(False)

ax2 = ax1.twinx()
ax2.plot(T,squeeze(W.getvalues()), color='#0A835B', linewidth=4.)
ax2.set_ylabel('Synaptic Weight', color='#0A835B', fontsize=20)
ax2.tick_params(labelsize=16)
ax2.set_xlim(xmax=end_time)
ax1.set_xlabel('time (s)', fontsize=20)
ax1.set_xticks(range(0,16,1))
```
