```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
T_trial = 1500 * ms
N_trials = 20
N_GR_groups = 3
N_GR_per_group = 3
N_GR = N_GR_groups*N_GR_per_group
N_MLI = 10
MLI_trg = 5

def pf_rates(t):
    '''
    returns a firing rate as a function of time. To be used with
    the Poisson group

    t: is the simulation time step
    '''
    t_ = t % T_trial
    fr = ones(3*N_GR_per_group) * 2*Hz
    if 400*ms <= t_ < 500*ms:
        fr[0:N_GR_per_group] = 200 * Hz
    elif 500*ms <= t_ < 600*ms:
        fr[N_GR_per_group:2*N_GR_per_group] = 200 * Hz
    elif 600*ms <= t_ < 700*ms:
        fr[2*N_GR_per_group:3*N_GR_per_group] = 200 * Hz
    return fr

def synapse_inds(neuron_inds, synapses, neuron_inds_position='pre'):
    '''
    takes a 'synapses' object and a set of pre/post synaptic neuron
    indices, returns a sorted array of post/pre synapse indices and
    correponding sorted array of pre/post synaptic neuron indices
    '''
    if neuron_inds_position == 'post':
        synapses_darr = synapses.synapses_post
    elif neuron_inds_position == 'pre':
        synapses_darr = synapses.synapses_pre
    else:
        raise Exception("Which must be 'pre' or 'post'")
    syns = array(synapses_darr)[neuron_inds]
    syns_inds = hstack(syns)
    neuron_inds = hstack(i*ones(s.shape[0],dtype=int) for i,s in enumerate(syns))
    sort_inds = argsort(syns_inds)
    return syns_inds[sort_inds], neuron_inds[sort_inds]

def calc_normal_firing_rates(rate_monitor, max_rate):
    '''
    calculate the normalized firing rates
    '''
    max_rate = float(max_rate)
    fr = rate_monitor.get_firing_rates()
    fr[fr>max_rate] = max_rate
    fr /= max_rate
    return fr    
```

```python
defaultclock.dt = .25*ms

MLI = MLIGroup(N_MLI)
S_MLI_MLI = Synapses(MLI,MLI,model='w:1',pre='g_inh+=MLI.g_inh_*w')
S_MLI_MLI.connect_random(sparseness=.4)
S_MLI_MLI.w[:,:] = 'rand()'

GR = PoissonGroup(N_GR,rates=pf_rates)
S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
S_GR_MLI[:,:] = '(j==5) | (rand()<.5)'
S_GR_MLI.w[:,:] = .15

CF = SpikeGeneratorGroup(1,[(0,(i*ms+j*T_trial)) for j in range(N_trials) for i in range(650,700,10)])
S_CF_MLI = Synapses(CF,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
S_CF_MLI[0,5] = 1.
S_CF_MLI.w[:,:] = 3 # lookup physio value

@network_operation(Clock(dt=defaultclock.dt))
def random_current():
    MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA

record_clock = Clock(1*ms)
GR_S = SpikeMonitor(GR)
GR_R = RealTimeRateMonitor(GR, record=True, record_clock=record_clock, tau_f=20*ms)
MLI_V = StateMonitor(MLI, 'V', record=True, clock=record_clock)
MLI_S = SpikeMonitor(MLI)
MLI_R = RealTimeRateMonitor(MLI, tau_f=100*ms, tau_r=20*ms, record=True, record_clock=record_clock)
W = StateMonitor(S_GR_MLI,'w',record=True, clock=record_clock)

beta = .005
syn_inds, pre_inds = synapse_inds(range(N_GR),S_GR_MLI,'pre')
_, post_inds = synapse_inds(range(N_MLI),S_GR_MLI,'post')
@network_operation(defaultclock)
def update_weights():    
    mli_fr = calc_normal_firing_rates(MLI_R, 100)
    gr_fr = calc_normal_firing_rates(GR_R, 500)    
    S_GR_MLI.w[:,:] += beta*gr_fr[pre_inds]*(mli_fr[post_inds] - S_GR_MLI.w[:,:])
        
run(N_trials*T_trial)
```
