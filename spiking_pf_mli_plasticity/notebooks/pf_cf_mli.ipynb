```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
N_GR_groups = 3
N_GR_per_group = 3
N_GR = N_GR_groups*N_GR_per_group

def pf_rates(t):
    '''
    returns a firing rate as a function of time. To be used with
    the Poisson group

    t: is the simulation time step
    '''
    fr = ones(3*N_GR_per_group) * 2*Hz
    if 400*ms <= t < 500*ms:
        fr[0:N_GR_per_group] = 100 * Hz
    elif 500*ms <= t < 600*ms:
        fr[N_GR_per_group:2*N_GR_per_group] = 100 * Hz
    elif 600*ms <= t < 700*ms:
        fr[2*N_GR_per_group:3*N_GR_per_group] = 100 * Hz
    return fr
```

```python
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    GR = PoissonGroup(N_GR,rates=pf_rates)
    MLI = MLIGroup(10)
    S_MLI_MLI = Synapses(MLI,MLI,model='w:1',pre='g_inh+=MLI.g_inh_*w')
    S_MLI_MLI.connect_random(sparseness=.4)
    S_MLI_MLI.w[:,:] = 'rand()'
    
    S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
    
    # STUPID BILL -- don't randomly initial weights every time!!!
    S_GR_MLI[:,:] = '(j==5) | (rand()<.5)'
    S_GR_MLI.w[:,:] = initial_weights

    CF = SpikeGeneratorGroup(1,[(0,i*ms) for i in range(650,700,10)])
    S_CF_MLI = Synapses(CF,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
    S_CF_MLI[0,5] = 1.
    S_CF_MLI.w[:,:] = 5 # lookup physio value

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        
    record_clock = Clock(1*ms)
    GR_S = SpikeMonitor(GR)
    GR_R = RealTimeRateMonitor(GR, record=True, record_clock=record_clock, tau_f=20*ms)
    MLI_V = StateMonitor(MLI, 'V', record=True, clock=record_clock)
    MLI_S = SpikeMonitor(MLI)
    MLI_R = RealTimeRateMonitor(MLI, tau_f=100*ms, tau_r=15*ms, record=True, record_clock=record_clock)
    W = StateMonitor(S_GR_MLI,'w',record=True, clock=record_clock)
    
    beta = .002
    @network_operation(defaultclock)
    def update_weights():
        mli_fr = MLI_R.get_firing_rates()
        mli_fr[mli_fr>100] = 100.
        mli_fr /= 100.
        gr_fr = GR_R.get_firing_rates()
        gr_fr[gr_fr>500.] = 500.
        gr_fr /= 500.
        delta_w = beta*gr_fr*(mli_fr[5] - S_GR_MLI.w[:,5])
        S_GR_MLI.w[:,5] += delta_w
        
    run(1000*ms)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,5]
```

```python
weights_history = [.15*ones(N_GR)]
MLI_history = []
for i in range(20):
    reinit()
    reinit_default_clock()
    clear(True)
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, final_weights = run_net(weights_history[-1])
    weights_history.append(final_weights)
    MLI_history.append(MLI_R.getvalues())
```
