```python
import sys
sys.path.append("../../")
from brian import *
from neuron_models import *
from spiking_pf_mli_plasticity import *
```

```python
def pf_rates(t):
    '''
    returns a firing rate as a function of time. To be used with
    the Poisson group

    t: is the simulation time step
    '''
    fr = zeros(3) * Hz
    if 400*ms <= t < 500*ms:
        fr[0] = 200 * Hz
    elif 500*ms <= t < 600*ms:
        fr[1] = 200 * Hz
    elif 600*ms <= t < 700*ms:
        fr[2] = 200 * Hz
    return fr
```

```python
def run_net(initial_weights):
    
    defaultclock.dt = .25*ms
    GR = PoissonGroup(3,rates=pf_rates)
    MLI = MLIGroup(1)
    CF = SpikeGeneratorGroup(1,[(0,i*ms) for i in range(650,700,10)])
    S_GR_MLI = Synapses(GR,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
    S_GR_MLI[:,:] = 1
    S_GR_MLI.w[:,:] = initial_weights
    S_CF_MLI = Synapses(CF,MLI,model='w:1',pre='g_ampa+=MLI.g_ampa_*w')
    S_CF_MLI.connect_one_to_one()
    S_CF_MLI.w[:,:] = 5

    @network_operation(Clock(dt=defaultclock.dt))
    def random_current():
        MLI.I = gamma(3.966333,0.006653,size=len(MLI)) * nA
        
    GR_S = SpikeMonitor(GR)
    GR_R = RealTimeRateMonitor(GR, record=True)
    MLI_V = StateMonitor(MLI, 'V', record=True)
    MLI_S = SpikeMonitor(MLI)
    MLI_R = RealTimeRateMonitor(MLI, tau_f=60*ms, tau_r=15*ms, record=True)
    W = StateMonitor(S_GR_MLI,'w',record=True)
    
    beta = .001
    @network_operation
    def update_weights():
        mli_fr = min(MLI_R.get_firing_rates(),100.)/100.
        gr_fr = GR_R.get_firing_rates()
        gr_fr[gr_fr>500.] = 500.
        gr_fr /= 500.
        delta_w = beta*gr_fr*(mli_fr - S_GR_MLI.w[:,:])
        S_GR_MLI.w[:,:] += delta_w
        
    run(1200*ms)
    
    return GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, S_GR_MLI.w[:,:]
```

```python
weights_history = [.35*ones(3)]
for i in range(50):
    reinit()
    reinit_default_clock()
    clear(True)
    GR_S, GR_R, MLI_V, MLI_S, MLI_R, W, final_weights = run_net(weights_history[-1])
    weights_history.append(final_weights)
```
