```python
MLI = MLIGroup(1)
CF = SpikeGeneratorGroup(1, spiketimes=[(0,200*ms),(0,300*ms)])
GR = SpikeGeneratorGroup(1, spiketimes=[(0,400*ms),(0,500*ms),])
model = '''w:1
           s:1'''
u,tau = .19, 2.1*second
S_CF_MLI = Synapses(CF,MLI,model, pre='''s+=(1-s)*(1-exp(-(t-lastupdate)/tau)); g_ampa_fast+=MLI.g_ampa_*s*w; 
                                         g_ampa_slow+=MLI.g_ampa_*s*w; n+=s*w; s*=u''')
S_CF_MLI.connect_one_to_one()
S_CF_MLI.w = .33
S_CF_MLI.s = 1.

S_GR_MLI = Synapses(GR,MLI,model='''w:1
                    v:1''',pre='g_ampa_fast+=MLI.g_ampa_*v; g_ampa_slow+=MLI.g_ampa_*v; n+=1')
S_GR_MLI[:,:] = 1
S_GR_MLI.w = .5
S_GR_MLI.v = .5

M_V = StateMonitor(MLI,'V',record=True)
M_ampa = StateMonitor(MLI,'g_ampa',record=True)
M_nmda = StateMonitor(MLI,'gv_nmda',record=True)
M_s = StateMonitor(S_CF_MLI,'s',record=True)
run(700*ms)
```
